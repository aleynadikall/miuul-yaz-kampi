```python
import warnings
import joblib
import pydotplus
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz, export_text
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate, validation_curve
from skompiler import skompile
import graphviz
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

# Olas baz warningleri grmezden gelmek iin:
warnings.simplefilter(action='ignore', category=Warning)
```

```python
################################################
# 1. Exploratory Data Analysis
################################################

################################################
# 2. Data Preprocessing & Feature Engineering
################################################

################################################
# 3. Modeling using CART
################################################

df = pd.read_csv("datasets/diabetes.csv")

y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)

cart_model = DecisionTreeClassifier(random_state=1).fit(X, y)
```

```python
# Confusion matrix iin y_pred:
y_pred = cart_model.predict(X)
```

```python
# AUC iin y_prob:
y_prob = cart_model.predict_proba(X)[:, 1]
```

```python
# Confusion matrix
print(classification_report(y, y_pred))
```

* %100 baarl bir sonu elde ettik. Bu teaorik olarak mmkn deildir. Belirli bir hata olmas lazm model kurmann temel esprisi gerei. Rastlantsal bir hata olmas lazm.

```python
# AUC
roc_auc_score(y, y_prob)
```

* Baar nasl 1 kabilir? Bu terste bir ilik yok mu? Var gibi gzkyor. Neden byle oldu peki? Az nce de ifade ettiimiz gibi sevinmeli miyiz yoksa zlmeli miyiz? Nasl yaklamalyz bir yere koyamyoruz. Buna ynelik olarak holdout yntemi grmtk. 

```python
#####################
# Holdout Yntemi ile Baar Deerlendirme
#####################

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=17)

cart_model = DecisionTreeClassifier(random_state=17).fit(X_train, y_train)
```

```python
# Train Hatas
y_pred = cart_model.predict(X_train)
y_prob = cart_model.predict_proba(X_train)[:, 1]
print(classification_report(y_train, y_pred))
roc_auc_score(y_train, y_prob)
```

```python
# Test Hatas
y_pred = cart_model.predict(X_test)

# Bu modele modelin hi grmedii test setini gnderdik:
y_prob = cart_model.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))
roc_auc_score(y_test, y_prob)
```
