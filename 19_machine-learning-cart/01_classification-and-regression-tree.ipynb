* 
* 
* 
* DİKKAT : Bir sınıflandırma problemi için Gini katsayısını ve entropiyi kullanabiliriz. Buradaki temel mantık şudur: Hatırlayalım bir yerlerde sıkıştığımızda gerçek değerlerle tahmin edilen değerler arasındaki farklara yöneldiğimizde bu muhtemelen bizi kurtarabilir şeklinde vurgular yapmıştık. Burada da benzer durumlar var aslında. Entropy değeri ve gini değerleri bize aslında tahminlerimizin başarısına yönelik bir bilgi sağlar. Elimizde gerçek sınıflar var. Bu sınıflara ilişkin tahminlerimiz var. Bu gerçek değerlerle tahminler arasındaki durumu değerlendirerek bize başarımıza ilişkin bilgi verir. Bu başarı metriklerine göre dallanmalara karar verilir. Regresyon problemini hatırlayalım. Orada SSE'yi bölme işlemi yapıp gerçek değerler ve tahmin edilen değerleri kıyasladıktan sonra sonra nihai olarak en düşük bölme noktasını şurası veriyor demiştik. Şimdi regresyon bölümünde kullandığımız SSE'ydi. Burada kullanacak olduklarımız ise sınıflandırma problemlerimiz için odak fonksiyonlarımız bunları da genellikle saflık ölçüleri denir. Yani benzerlik benzemezlik gibi düşünürsek ne kadar düşük o kadar iyidir. Entropy ve gni değerleri sınıflandırma problemleri için hatamızı değerlendirme imkanı sağlar. Mesela dallara ayırdıktan sonra dalların sağ tarafı ve dalların sol tarafı için ağırlıklı bir şekilde kaybımız / lossumuz / costumuz / hatamız hesaplanıyor olur.

* Entropy birçok farklı disiplinde kullanılan bir değerdir. Bizim istatiki ve şu anda ilgilendiğimiz kısmı ise ne kadar düşük entropi o kadar iyidir. Çünkü entropi çeşitlilik / varyans / bilgi demektir. Şimdi tahmin edeceğimiz değerlere odaklanacak olduğumuzdan dolayı bunların çeşitli olmamasına dikkat ederiz. Bunların düşük olmasını bekleriz. Mesela bir dalda/yapraktayız. Bu yaprak içerisinde çeşitli tahminlerimiz var. İki sınıfımız olduğunu düşünelim. Bir sınıfının gerçekleşmesi olasılığı verilmiş. Çarpı. Bu gerçekleşme olasılığının logaritması verilmiş. Bunu sınıflandırma modeli başarı metriklerini değerlendirirken görmüştük. Buradan çıkacak olan değer aslında 1 sınıfına sahip olma olasılığı ve 0 sınıfına sahip olma olasılıklarını taşıdığından dolayı bu değerler ne kadar küçük olursa o kadar iyi oluyordu. Yani örneğin bir yaprak içerisinde bütün değerler 1 ise 0 hiç yok ise bu durumda k=1'den K'ya kadar 1.sınıfın tahmin olasılığı buradan gelecek. 0.sınıf hiç yok, 0 sınıfı hiç olmadığı için olabilecek en düşük entropi gelecektir bize. Diyelim ki burada 0'lar da var yani çeşitlilik var dolayısıyla 1 sınıfının gerçekleşmesi olasılıklarına ilişkin bir hesap seti var artı 0 sınıfının olasılıklarına ilişkin bir hesap seti var. Bunların hepsini topladığımızda daha yüksek bir değer gelecektir. Dolayısıyla entropi ne kadar düşükse o kadar iyidir.

```python
import warnings
import joblib
import pydotplus
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz, export_text
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate, validation_curve
from skompiler import skompile
import graphviz
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

# Olası bazı warningleri görmezden gelmek için:
warnings.simplefilter(action='ignore', category=Warning)
```

```python
################################################
# 1. Exploratory Data Analysis
################################################

################################################
# 2. Data Preprocessing & Feature Engineering
################################################

################################################
# 3. Modeling using CART
################################################

df = pd.read_csv("datasets/diabetes.csv")

y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)

cart_model = DecisionTreeClassifier(random_state=1).fit(X, y)
```

```python
# Confusion matrix için y_pred:
y_pred = cart_model.predict(X)
```

```python
# AUC için y_prob:
y_prob = cart_model.predict_proba(X)[:, 1]
```

```python
# Confusion matrix
print(classification_report(y, y_pred))
```

* %100 başarılı bir sonuç elde ettik. Bu teaorik olarak mümkün değildir. Belirli bir hata olması lazım model kurmanın temel esprisi gereği. Rastlantısal bir hata olması lazım.

```python
# AUC
roc_auc_score(y, y_prob)
```

* Başarı nasıl 1 çıkabilir? Bu terste bir işlik yok mu? Var gibi gözüküyor. Neden böyle oldu peki? Az önce de ifade ettiğimiz gibi sevinmeli miyiz yoksa üzülmeli miyiz? Nasıl yaklaşmalıyız bir yere koyamıyoruz. Acaba overfit'e mi düştük yoksa gerçekten başarılı mıyız bilemiyoruz. Dolayısıyla sorulması gereken bir soru var o da şu: Bu başarımızı ya da hatalarımızı nasıl daha doğru değerlendirebiliriz? Buna yönelik olarak holdout yöntemi görmüştük. 

```python
#####################
# Holdout Yöntemi ile Başarı Değerlendirme
#####################

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=17)

cart_model = DecisionTreeClassifier(random_state=17).fit(X_train, y_train)
```

```python
# Train Hatası
y_pred = cart_model.predict(X_train)
y_prob = cart_model.predict_proba(X_train)[:, 1]
print(classification_report(y_train, y_pred))
roc_auc_score(y_train, y_prob)
```

```python
# Test Hatası
y_pred = cart_model.predict(X_test)

# Bu modele modelin hiç görmediği test setini gönderdik:
y_prob = cart_model.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))
roc_auc_score(y_test, y_prob)
```

* Gözlemliyoruz ki model eğitildiği veride performansını çok yüksek gösterdi ama bilmediği bir veriyi sorduğumuzda precision değeri, recall değeri, f1 skoru nerdeyse yarı yarıya fark etti. Anlaşılacağı üzere bu terslikte bir iş var. Model görmediği veride oldukça kötü bir sonuç üretti. Çünkü train setini çok iyi öğrendi, ezberledi. Bundan dolayı hiç görmediği bir veriyi sorduğumuzda başarısı düştü. Hem de çok ciddi bir şekilde düştü. %67 değeri kayda değer bir değer mi accuracy açısından bakacak olursak kayda değer olabilir iyi mi değil kötü mü değil, bir şekilde kayda değer olabilir. Fakat şimdi başka bir şüphemiz / kaygımız var. Burada bu veri setini bölerken 30'a 70 böldük. Random_state'ini de aynı sonuçları elde etmek için 17 yaptık. Meslea random_state'i 45 yapsaydık:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=45)

cart_model = DecisionTreeClassifier(random_state=17).fit(X_train, y_train)
```

```python
# Train Hatası
y_pred = cart_model.predict(X_train)
y_prob = cart_model.predict_proba(X_train)[:, 1]
print(classification_report(y_train, y_pred))
```

```python
roc_auc_score(y_train, y_prob)
```

```python
# Test Hatası
y_pred = cart_model.predict(X_test)

# Bu modele modelin hiç görmediği test setini gönderdik:
y_prob = cart_model.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))
roc_auc_score(y_test, y_prob)
```

* Yine oldukça kayda değer, ilginç sonuçlar geldi. AUC 65 oldu. Daha önce 67'ydi. Accuracy'imiz de 0.71'den 0.69'a düştü. Bir türlü içerisinden çıkamadık. Ama bir şeyi gözlemledik. Model zaten overfit etti. Train setinde çok yüksek test setinde çok düşük çıkmasının sebebi budur. Zaten train ve test setinde fark olmasını bekleriz. Fakat burada olağan üstü bir fark vardır. Modelimiz veri setini overfit etmiş, ezberlemiştir. Dolayısıyla test ve train ayrımından da bir şüphe ediyoruz. Lehimize ya da aleyhimize başarı açısından başarı metrikleri değişebiliyor. Bu durumda çapraz doğrulamaya gitmemiz gerekiyor. 

```python
#####################
# CV ile Başarı Değerlendirme
#####################

cart_model = DecisionTreeClassifier(random_state=17).fit(X, y)
# Buradaki fit'i koymayabilirdik de. Çünkü burada şöyle bir soru olabilir. Daha önce yaptığımız gibi cross validation'ı 
# getireceğiz. Zaten burada model fit edilmişti biz aslında sağlıklı bir şekilde değerlendiremiyoruz ki çünkü bğtün veriye fit 
# edildi zaten buradaki veri gibi bir sorunumuz olabilir. Bu geçerli doğru bir sorudur. Fakat cross validate zaten bu fit 
# metodundan gelen özellikleri görmezden geliyor. Burada kendisi zaten parçalara böldükten sonra fit etme ve başarı 
# değerlendirme işlemini kendisi yapıyor. Dolayısıyla bunu burada yazmayabiliriz de. 

cv_results = cross_validate(cart_model,
                            X, y,
                            cv=5,
                            scoring=["accuracy", "f1", "roc_auc"])

cv_results['test_accuracy'].mean()
```

```python
cv_results['test_f1'].mean()
```

```python
cv_results['test_roc_auc'].mean()
```

Normalde veri seti yeteri kadar zengin olduğunda sadece direkt train test ayrımı da tercih edilebilir. Ya da başka kıyaslamalar yapılacak olduğunda yani yeni hiperparametreler eklendiğinde ya da yeni değişkenler eklendiğinde bir hataya göre ilerlemek istediğimizde tek başına yine train test ayrımı da kullanılabilir. Yani şu anda bu elde ettiğimiz hatalar en doğru en geçerli hatalardır/başarılardır. Diğer almış olduğumuz hataların bazılarının üstünde bazılarının altında. ###### Şu an sonuçlar da sanki biraz düşük oldu, model başarımızı nasıl arttırabiliriz?

* Model başarımızı yeni gözlemler ekleyerek, yeni değişkenler ekleyerek, veri ön işleme işlemlerine dokunarak yada hiperparametre optimizasyonu yaparak gerççekleştirebiliriz. Ek olarak bu problem için dengesiz veri yaklaşımları da başarımızı arttırmaya yardımcı olabilir. Dengesiz veri yaklaşımları temelinde gözlem birimlerinin dallarının yani iki sınıf var diyelim bunların oranları birbirinden çok farklı diyelim bağımlı değişkendeki sınıfların dağılımları bunları birbirine azaltarak, arttırarak, rastgele örneklem yöntemi gibi yöntemler kullanarak yaklaştırmaya çalıştığımız, buradaki dengesizliğin giderilmeye çalışıldığı yöntemlerdir.
