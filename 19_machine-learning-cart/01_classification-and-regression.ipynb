### CART ( Classification & Regression Tree )

* Karar ağaçları makine öğrenimi dünyasında çok önemli bir yere ve çok geniş bir kullanım alanına sahiptir. Bu yöntemlerin gelişmiş yöntemleri hali hazırdaki birçok makine öğrenmesi yarışmasında en yüksek performanslı yöntemlerdir. Tahmin başarısı açısından, dolayısıyla bu yöntemin temelini oluşturan CART bizim için çok öenmlidir. CART'ı oldukça detaylı bir şekilde teorisini ve uygulamasını gerçekleştirerek öğrenmiş olacağız ve kullanılacak olan ileri seviye olarak kabul edebileceğimiz birçok yöntemin de temellerini oluşturduğundan dolayı karar ağaçları yöntemlerine sağlam bir giriş yapmış olacağız.

* 

* CART RandomForest'ın temelini oluşturur.

* 

* Bir excel düşünelim, excelde filtreleme işlemi yaparak nihayetinde sayısal bir değişkenin ortalamasını aldığımızı ya da kategorik bir değişkenin frekans bilgisini aldığımızı düşünelim. Tam olarak ana işleyiş bu şekildedir. Fakat burada değişkenleri neye göre filtreleyeceğimiz neye göre gruplara ayıracağız sorunu programatik/algoritmik bir şekilde çözülmüş ve karar ağaçları kullanılıyormuş gibi düşünebiliriz.

* 

* Bağımlı değişken maaş, bağımsız değişkenler deneyim yılı ve atış sayısıdır. Bir kara ağacı şemasında en tepede görülen değişken en önemli değişkendir. Dolayısıyla buradaki en önemli değişken deneyim yılı ondan sonra da atış sayısıdır. Bağımsız değişkenler karşımıza açıklayıcı değişken veya feature olarak da gelebilir.

* Bağımszı değişkenlerin dallandığı/bölündüğü noktalara iç düğüm noktaları/bölüm denir. Bu ağacın 2 tane iç düğüm noktası vardır. 3 tane de son düğüm noktası/yaprakları vardır. Bu değerler ise exceldeki gibi filtre işlemi yapıldıktan sonra ortalaması alınmıştır. Regresyon problemlerinde bu dallanma işlemlerinden sonra kalan bölgelerin ortalaması alındığında bunu nihai tahmin sonucu olarak verilir.

* 

* Aslında bir karar ağacı kullandığımızda nihaiyetinde bu yöntemler bize bir karar kuralı çıkarır. Predictor B ve Predictor A bağımsız değişkendir ve bunların kesişiminde bağımlı değişken vardır. 3 bölümdeki sayılar da (2.5, 1.3, 5,6) ortalamadır.

* 

* Karar ağaçları temelinde çalışmanın başındaki heterojen veri setlerini bu şekilde karar kurallarına bölerek tahmin sonuçları elde eder ve ifade etmiştik veri bilimi ve makine öğrenmesi dünyasında tabular verilerde en başarılı olan yöntemler ağaç yöntemlerdir. Görüntü işleme metin işleme problmeleri gibi problemlerde derin öğrenme temelli yöntemler ağaç yöntemlerine göre daha iyidir.

#### Bu işlemi excelde neden yapmıyoruz?

* x1 değişkenini neye göre böleceğiz bunu excelde bilemeyiz. Biz burada bir algoritmanın çalışma sonucunu görüyoruz. Karar kurallarını çıkarmak excelde mğmkün değildir. Karar kyrallarını çıkarttıktan sonra bunu excele dökmek mümkündür. Algoritmik bir şekilde belirli referans noktalarına göre hareket ederek kendimiz optimize ederek bu karar kurallarının ne olduğunu belirlemektir. Zaten işin ana noktası bu karar noktalarını bulmaktır. Bunu excelde bu yüzden yapamayız. Karar kurallarını belirlendikten sonra excel de dahil çoğu ortamda çalıştırılabilir hale gelir.

* 

* DİKKAT : Bu ağacı nasıl oluşturmalıyız sorusunun yanıtı RSS diğer ifadesiyle SSE değerinin minimum olduğu noktalardaki bölgelenmeler yapraklanmalar kutulanmalardır. Bu ne demek? Şimdi gerçek değerler yi tahmin edilen değerlerin yanında yj yazmıyor Rj ifadesi yazıyor. Bu Rj'ler bölgeleri ifade ediyor. Önemli bir farkındalık noktasındayız. Regresyon problemleri için karar ağaçları kullandığımızda tahmin edilen değerler bölme işlemleri neticesinde ortaya çıkan kutularda kalan gözlem birimlerinin ortalamasıdır. Dolayısıyla bir karar noktasında bölme/dallanma yaptık. Kutunun içerisindeki kalan gerçek değerlerin ortalaması alınır bunlar tahmin sonucu olarak değerlendirilir. Dolayısıyla gerçek değerlerle tahmin edilen değerlerin farkı dediğimizde buradaki tahmin edilen değerler (yRj) ilgili kutuların ortalama değerleridir. İşte biz bu dallanmaları/budaklanmaları öyle bir şekilde yapmalıyız ki öyle bir noktada kapatmalıyız veya derinleştirmeliyiz ki bu toplam hatalar minimum olsun.
