### Encoding
* Değişkenlerin temsil şekillerinin değiştirilmesi...
* Göreceğimiz yöntemler:
* 1 - Label Encoding: Bir kategorik değişkenin sınıfları labellardır. Buradaki labelları yeniden kodlamak demektir. 
* GÖRSEL 1: Kategorik değişken sınıfları


* GÖRSEL 2: Kategorik değişken sınıflarının encoding edilmiş hali


* Başka bir senaryoda bunu değerlendirelim:
* Elimizde eğitimle ilgili bir değişken var ve bu eğitim durumu değişkeni label encoding ile aşağıdaki şekilde encoding edilmiş olsun.

* Ordinal bir kategorik değişken...
* DİKKAT: Bu değişken label encoderdan geçirdik ve 0 1 2 3 4 5 diye dönüştürdük diyelim, bu kategorik değişkenin numerik temsilleri bunlar oldu diyelim. Evet bu değişkeni bu şekilde dönüştürebilir miyiz? Evetse neden, hayırsa neden? 
* Evet dönüştürebiliriz. Neden dönüştürebiliriz çünkü buradaki education kategorik değişkeni sıralıdır (ordinaldir). Eee bu ordinal kategorik değişkeninin sınıfları arasında fark vardır durumu bu değişkenin sayısal olarak da temsil edildiği durumda taşınmaktadır yani 0 burada küçük sınıfı 5 burada büyük sınıfı temsil etmektedir. Dolayısıyla bu education label encoderdan bu şekilde geçirilebilir. Yani sorumuzun yanıtı eveet geçirilebilir. Bir diğer yanıt şöyle olabilirdi. Evet geçirilebilir ama one hot encoderdan da geçirilebilir olabilirdi. Bu da alternatif bir yanıt olabilirdi. Ancak soru kapsamı sadece bunun bu şekilde dönüştürülüp dönüştürülemeyeceğiyle ilgiliydi. 

* Diğer bir senaryo:

* DİKKAT: Buradaki bu değişken label encoderdan bu şekilde geçirilebilir mi? Evetse neden, değilse neden? Eğer bu şekilde kalacaksa yapılmamalıdır. Çünkü bu durumda ilgili yöntemler yanikullanacak olduğumuz modelleme teknikleri bu sayısal değerleri büyüklük olarak algılayacaktır. Normalde takımlar arasında herhangi bir farklılık yok. Bunlar veri setinde gözlemlenen takımlar. Bu takımların arasında bir fark olmadığı için bir ordinallik/sıralı durum olmadığından dolayı label encoderdan nominal bir kategorik değişkeni geçirmek bir miktar sakıncalı olacaktır. Kullanılacak algoritmalara göre farklılık gösteriyor olsa da buradaki büyüklük küçüklük durumu normalde olmadığı halde sanki ilk takımın değeri son takımın değeri büyükmüş gibi olmaktadır. Bundan dolayı label encoding böyle bir değişkende böyle bir şekilde kullanılmamalıdır. Çünkü öncesinde var olan büyüklük küçüklük algısı burada bozulmuş olacaktır. Peki bu şekilde dönüştürülen bir değişkeni daha doğru nasıl dönüştürebiliriz? Burada yapılması gereken işlem one hot encoderdan geçirmektir. One hot encoding buradaki her bir sınıfı sütun olarak değişkene dönüştürülecektir.
