### Rare Encoding

* Rare nadir demektir. Az gözlemlenen demektir. Bunu bir örnek üzerinden birlikte değerlendirelim. Diyelim ki elimizde aşağıdaki gibi bir değişken var.

* Bu değişken kategorik değişken ve bu değişkenin sınıfları var. Bu sınıfların tüm verideki gözlenme frekansları verilmiş yani diyelimki elimizde 1 milyon satırlık ya da 100 bin satırlık veri olsun, bu veri içerisinde de bu kategorik değişkenin çeşitli sınıfları olsun. Bunlar çeşitli şehirler gibi düşünelim. Bu şehirlerin gözlenme frekansları 56, 84, 54, 2, 12, 60, 3, ... şeklinde. Burada birçok yüksek değer varken düşük değerler de var. Mesela diyelim ki bu düşük değerleri ne yapacağız problemimiz var. İyi de kalsın öyle bırakalım ne işimiz var bunlarla diyor olabiliriz.
* DİKKAT: Genelde model geliştirme süreçlerinde karmaşıklık ile değil basitlik ve genellenebilirlik ile ilgileniyor oluruz. Buradaki genellenebilirlik biz herkesi kapsayalım değil, büyük çoğunluğu temsil edelim gibi düşünülebilir. Bu şu anlama gelir: Bir model kurucaz ve bu modeli çeşitli sistemlere entegre ediyor olacağız. Örneğin ev/araç fiyat tahmini ya da buna benzer bir iş dalında benzer bir problemi çözecek olan bir makine öğrenmesi olsun. Bizim yüz binlerce kaydımızın içerisinde sadece 2 kere gözükmüş D sınıfı varya biz bunu one_hot_encoder'dan geçirdiğimizde ne olacak bu bir değişkene dönüşmüş olacak. Bu değişkenin sadece iki tane hücresinde 1 diğer on binlerce yüz binlerce hücresinde 0 yazmış olacak. Bu bir bilgi taşıyor diyebilir miyiz? Ayırt edicilik sağlar diyebilir miyiz? Diyemeyiz. Çok az sayıda gözleme sahip  yani şu anlama geliyor elimizdeki örnek veride yüz binde iki tane olan bir şey bütün model kurduğumda da bu frekansta olmasını bekleyeceğimizden zaten modele bu değişkenden değer gelme ihtimali çok çok düşük benzer şekilde burada ki bir başka sınıf olan G verimizde 3 defa gözlemlenmiş, biz bunu bir değişkene çevirsek bunun üzerinden bir tahmin işlemi yapacak olsak zaten bunun da gözlenme frekansı çok çok düşük yani bunun modelleme anlamında karşılık bulma ihtimali çok düşük diğer yandan taşıdığı bilgi anlamında bağımlı değişkene yönelik bir ayırt ediciliğe sahip olup olmaması da tartışılır yani aslında bu 2, 3, 5 tanenin hepsi örneğin bir sınıflandırma problemindeki ilgili bir sınıfa biel ait olsa yine çok az rastgele çıkmış olabilir bilemiyoruz dolayısıyla bu kategorik değişkenin sınıflarını one hot encoderdan geçirip burada bazı yeni değişkenler oluşturacağız ya bu oluşturduğumuz değişkenlerin de ölçüm kalitesi olsun istiyoruz bu oluşturacağımız değişkenlerin de bağımlı değişkene çeşitli etkileri olma ihtimalinden gitmek istiyoruz bu sebeple ve gereksiz birçok fazla değişken oluşturmak istemediğimizden dolayı gereksiz birçok fazla değişken hem iterasyon süreçlerine hem optimizasyon süreçlerine hem de örneğin ağaç yöntemlerdeki bölünme, dallanma süreçlerini çok ciddi etkiliyor olacaktır. Bu sebeple gereksiz değişkenlerden uzaklaşmak/kurtulmak çabamızın da bir neticesi olarak rare encoder'ı kullanabiliriz.
