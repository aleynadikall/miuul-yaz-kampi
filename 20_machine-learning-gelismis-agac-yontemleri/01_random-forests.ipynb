### Ratgele Ormanlar (Random Forests)

* 

###### Random Forest'a Neden Random Deriz?

* Çünkü makine öğrenmesi temel kaygıları olan genellenebilirlik, aşırı öğrenme gibi konularda da farkındalık olması istenir. Bu sebeple de mülakatlarda sorulabilir. 

* 

* DİKKAT : 2001 yılı yine Breiman CART yöntemini geliştirmişti. CART yöntemi oldukça iyi öğrenme kapasitesi olan bir yöntemdir. Aşırı öğrenmeye meyilli bir yönttemdir. Breiman bu aşırı öğrenmeye olan meyili çözmek için CART yönteminde dallanmayı belirli bir noktada kesmek için bazı yaklaşımlar geliştirmiştir. Yani ağacı budama işlemleri gerçekleştirmiştir. Bu bir şekilde işini görüyor olsa da yine de bir şekilde başka arayışlarını sürdürüyor olmuş. İşte bu arayışların sonucunda yine kendisinin çıkardığı bir yöntem olan Bagging ve diğer bir yöntem olan Random Subspace yöntemlerini kullanarak Rndom Forest yöntemini ortaya çıkarmış. Şimdi buradaki temel çalışma prensibi nedir? Daha önce biz bir tane CART gördük mesela bunun gibi örneğin 3, 10, 5, 500 tane ağaç olduğunu düşünün her bir ağaca fkri sorulduğunun ve her bir ağacın belirli bir sınıflandırma veya regresyon problemi için fikrini ifade ettiğini düşünün. Bu fikirkerin bir araya getirilerek nihai karara varıldığını düşünün. Yani tıpkı bir meclis gibi tıpkı bir çoğulcu karar alma mekanizması gibi birden fazla ağacın bir araya gelerek tahminler yaptığı yöntemlerden birisidir. Topluluk öğrenme modellerinin de aslında bu şekilde girişinde bulunuyoruz. 

* 
* Random Forest rastgeleliğini nereden alıyor denildiğinde aklımıza iki şey gelmelidir.
* 1 - gözlem birimlerinden ratgele seçerek n-tane örneğin ağaç oluşturur.
* 2 - Oluşturduğu ağaçlarda bölme işlemlerine başlayacak ya bu CART'ta gördüğümğx gibi bu bölme işlemine başlamadan önce değişkenlerşn içerisinden daha az sayıda rastgele değişkenler seçer. Bu seçilen değişkenlere bakıp bunlara göre dallanma işlemlerine karar verilir. Bu şekilde aslında değişken seçimi ve gözlem seçimiyle rassallığı korumuş olur, genellenebilirliği korumaya çalışır. Random Forest ismini buradan almaktadır.

* 

* 

* Breimann bagging yönteminde şöyle bir yol izliyor: CART da tek bir ağaç vardı. Örneğin onlarca ağaç fit eder onlarca fit ettiği ağaca fikirlerini sorarak tahminler elde ediyordu. Yani tek başına olan bir CART algoritmasını aslında çoğaltarak baggibg yöntemini ortaya çıkarmıştır. Bagging yönteminde elde ettiği bu rastgelelik çünkü örneğin 1000 tane gözlem birimi varsa bunun içerisinden örneğin 750 tane gözlem birimi seçilip bunun üzerine bir ağaç kuruluyor. Dolayısıyla bu rastgelelik ile gözlemlemiş olabileceği başarı ile madem ben gözlem seçimlerinde rastgele bir seçim işlemi yaptım ve buradan bir rassallık, tahmin başarısı yakaladım diye düşünülmüş olmalı ki daha sonra random subspace yöntemini de kullanarak yani gözlemlerde rastgele seçim yaptım yeni ağaçlar kurdum tamam başarılı ee demek ki random subspace yöntemiyle değişkenlerde de rastgele seçimler yapsak bu da başarılı olur gibi düşünülmüş olmalı ki random subspace yöntemini de bir araya getirmiş ve gözlemlerde ve değişkenlerde rastgelelik kavramını getirerek rastgele ormanlar yöntemini ortaya çıkarmış. Burada önemli bir nokta vardır. 

* 

* Bu rastgelelik kavramı genellenebilirlik kavramı yani CART'ı kullanıyorduk şimdi niye kaşındık ki yeni yöntemler arıyoruz, CART çok başarılı ama aşırı öğrenmeye meyilli dolayısıyla öyle bir işlem yapılmalı ki rastgelelik korunabiliyor olmalı, tahmin başarısı rastgeleliği koruyarak arttırılabiliyor olmalı. Dolayısıyla bu arayışların sonucu olarak önce bagging ortaya çıkmış ve baggingle beraber breiman random subspace'i birleştirerek gözlem seçimlerinde rastgelelik ve değişken seçimlerinde rastgeleliği getirerek random forestı ortaya çıkarıyor. 

* 

* Yukarıdaki yapı random forestın temelini oluşturan bir yapıdır. Buradaki yapının kavranması oldukça önemli çünkü bagging yöntemi yani topluluk öğrenme yöntemleri birbirinden bağımsız birçok ağaç fit etme yaklaşımı ağaç yöntemlerinde bir dönümnoktası ve mihenk taşıdır. Dolayısıyla ağaç yöntemleri incelendiğinde literatürün temel modellerden sonra bagging yöntemleri ve boosting yöntemleri olarak ikiye ayrıldığını görürüz. Bagging yöntemi demek aslında herhangi bir tekli ağaçlı bootstrap örnekleme yöntemiyle tekrar tekrar oluşturarak, rastgelelik kazandırarak tahminlerde bulunmak demektir. CART'a bunu uyguladığımızda bag edilmiş CART deriz. Dolayısıyla bagging asında bir metodoloji, birçok farklı ağaç yöntemine uygulsnsbilir. İşte bunun şimdi temel çalışma mantığını öğreneceğiz.

* Diyelim ki elimizde n adet gözlem birimi var bu n adet gözlem birimi içerisinden n'er adet ve n'den daha küçük sayıda bootstrap örnekleme yöntemiyle ağaçlar oluşturulur. Diyelim ki veri setimizde 1000 tane gözlem birimi var. Bu veri seti içerisinden rastgele 750 tane gözlem seçiliyor olsun. Bununla bir tane ağaç modeli kurulur. Daha sonra bu 750 gözlem birimi varya bu yerine konmuş gibi düşünülür. Buradan yeni bir tane 750 seçilir. Yeniden seçilen 750 gözlem birimi ile yeni bir ağaç modeli kurulur. Bu şekilde yerine koymalı ve gözlem biriminden daha az sayıda olacak şekilde yeni ağaçlar fit edilir ve burada kurulmuş olan T adet karar ağacımodelinin ürettiği T adet tahmin değeri bir araya getirilerek değenlendirilir. Bu bagging yöntemidir, random forestın temelini oluşturmaktadır. Bagging yönteminin kilit noktası bootstrap rastgele örneklem yöntemidir. Burada oluşturulan ağaçların önceki ağaçlar ile herhangi bir bağımlılığı yoktur. Örneklerin her bir ağaç için oluşturulma sürecindeki kullanılma şansı eşittir. Ağaçlar birbirinden bağımsızdır. Bagging yöntemi ve boosting yöntemi arasındaki fark bagging yöntemlerinde ağaçların birbirine bağımlılıkları yoktur, boosting yöntemlerde ise ağaçlar artıklar üzerine kurulur. Dolayısıyla ağaçların birbirilerine bağımlılıkları vardır. Mülakatta sorulabilir. Bu bagging yöntemi daha önce görmüş olduğumuz CART ve birçok tekli ağaç yöntemine göre hata kareler ortalamasını düşüren ya da doğru sınıflandırma başarısını arttıran ezberlemeye karşı dayanıklı bir yöntemdir. Anlaşılıyor ki bagging yöntemi random forestın temelindedir. Peki bu random subspace ne demek? Diyelim ki veri setinde 100 adet değişken olsun. Hani baggingde bir ağaç seçmiştik ya, sonra bir tane sonra bir tane daha.
```python
################################################
# Random Forests, GBM, XGBoost, LightGBM, CatBoost
################################################

import warnings
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.model_selection import GridSearchCV, cross_validate, RandomizedSearchCV, validation_curve

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# !pip install catboost
# !pip install xgboost
# !pip install lightgbm
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

warnings.simplefilter(action='ignore', category=Warning)

df = pd.read_csv("datasets/diabetes.csv")

y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
```

```python
################################################
# Random Forests
################################################

rf_model = RandomForestClassifier(random_state=17)
rf_model.get_params()
```

* Buradaki n_estimators fit edilecek bağımsız ağaç sayısını ifade etmektedir. Dolayısıyla görüyoruz ki bunun ön tanımlı 100 am belki 500 ağaç olduğunda daha iyi çalışacak. Bilmiyoruz ki. Diğer değerler için de aynı şekildedir. Bu nedenle bunların en iyi değerlerini gridsearch ile aramaya sokacağız.

```python
# Çalışmanın başındaki (hiperparametre optimizasyonu öncesindeki) hata değerleri:
cv_results = cross_validate(rf_model, X, y, cv=10, scoring=["accuracy", "f1", "roc_auc"])
cv_results['test_accuracy'].mean()
```

```python
cv_results['test_f1'].mean()
```

```python
cv_results['test_roc_auc'].mean()
```

```python
# Hiperparametre setimizi girelim:
rf_params = {"max_depth": [5, 8, None],
             "max_features": [3, 5, 7, "auto"],
             "min_samples_split": [2, 5, 8, 15, 20],
             "n_estimators": [100, 200, 500]}


# GridSearchCV yöntemi ile gezelim:
rf_best_grid = GridSearchCV(rf_model, rf_params, cv=5, n_jobs=-1, verbose=True).fit(X, y)

# En iyi parametreleri elde etme:
rf_best_grid.best_params_
```

* DİKKAT : Diyelim ki hiperparametre optimizasyonu yaptıktan sonra başta elde ettiğimiz değerden daha düşük bir hata çıktı. Normalde şunu bekleriz: Hiperparametre optimizasyonu sonrası elde edilen hatanın daha düşük olmasını bekleriz. Düşük olmamasının sebeplerinden birisi rastgelelik ile ilgili olabilir. Bir diğeri ön tanımlı argümanları arama setimizde bulundurmuyor olmamız olabilir. Dolayısıyla max_depth'in ön tanımlı değeri None ve max_features ön tanımlı değeri otomatik, min_sample_split ve n_estimators'ın da tanımlı değerleri burada barındırıyoruz.

```python
# rf_model'e set_params'ı kullanarak bu parametreleri set edelim:
rf_final = rf_model.set_params(**rf_best_grid.best_params_, random_state=17).fit(X, y)

cv_results = cross_validate(rf_final, X, y, cv=10, scoring=["accuracy", "f1", "roc_auc"])
cv_results['test_accuracy'].mean()
```

```python
cv_results['test_f1'].mean()
```

```python
cv_results['test_roc_auc'].mean()
```

```python
# Feature impportance:
def plot_importance(model, features, num=len(X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                     ascending=False)[0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')

plot_importance(rf_final, X)

# Tüm değişkenler ve önem düzeyleri geldi: 
```

```python
def val_curve_params(model, X, y, param_name, param_range, scoring="roc_auc", cv=10):
    train_score, test_score = validation_curve(
        model, X=X, y=y, param_name=param_name, param_range=param_range, scoring=scoring, cv=cv)

    mean_train_score = np.mean(train_score, axis=1)
    mean_test_score = np.mean(test_score, axis=1)

    plt.plot(param_range, mean_train_score,
             label="Training Score", color='b')

    plt.plot(param_range, mean_test_score,
             label="Validation Score", color='g')

    plt.title(f"Validation Curve for {type(model).__name__}")
    plt.xlabel(f"Number of {param_name}")
    plt.ylabel(f"{scoring}")
    plt.tight_layout()
    plt.legend(loc='best')
    plt.show(block=True)

val_curve_params(rf_final, X, y, "max_depth", range(1, 11), scoring="roc_auc")
```
