### Outliers (Aykırı Değer)

* Verideki genel eğiliminin oldukça dışına çıkan değerlere aykırı değer denir.
* Aykırı deeğerler ne gibi problemlere sebep olur bir görselle inceleyelim:


* Soldaki garfikte herhangi bir aykırı değer yok. x ile y arasındaki ilişkiyi bir doğru aracılığıyla modellenmiş, yani bu ki değişkenin aralarında bir ilişki var, yönleri ve şiddetleri de görseldeki şekildedir. Bu veri setine 3 tane aykırı değer eklendiğinde ise bu iki değişken arasındaki ilişkiyi ifade eden doğru nasıl değişiyor. Yönü ve şiddeti de değişmiş. Normalde negatif bir ilişki tanımı varken ilişki pozitife dönüşmüştür.

* DİKKAT: Dolayısıyla özellikle doğrusal problemlerde aykırı değerlerin etkileri daha şiddetlidir.Ağaç yöntemlerinde bu etkiler daha düşüktür.

* Özetle aykırı değerler bizim için dikkat edilmesi gereken bir husustur.

##### Peki aykırı değerleri neye göre belirliyoruz?

* 1 - Sektör Bilgisi
* 2 - Standart Sapma Yaklaşımı
* 3 - Z-Skoru Yaklaşımı
* 4 - Boxplot (Interquartile range - IQR) Yöntemi (Biz yaygınca tercih edilen tek değişkenli olarak boxplot yöntemini, çok değişkenli olarak da LOF yöntemini birlikte değerlendireceğiz.)


* Bu Yöntemin Çalışması:


* DİKKAT: Diyelim ki elimizde yaş adında bir değişken var bu yaş değişkeniyle ilgili aykırı değer hesabı yapmak istiyoruz. Bir boxplot çizdirdiğimizde outliersları yukarıdaki şekilde (bize göre solda kalıyor...) görebiliriz. Maksimum ve minimum değerleri de görüyoruz. Medyan değerimiz ise Q2' dir. Q1 %25'e Q3 %75'e denk gelen çeyreklerdir.

* Bu yaş değişkenini küçükten büyüğe sıraladığımızı düşünelim. Örneğin bu sıralama sonucunda en küçük yaş 2 olsun, Q1 değerimiz 20, Q2 değerimiz 30 olsun. Bu şekilde küçükten büyüğe doğru sıralandığında yüzdelik olarak en ortadaki mi, 75'lik çeyrek değerde mi şeklindeki soruların yanıtları olan değerleedir buradaki değerler.

* Aslında bir sapma, değişim veya aralıktır. Neyin aralığıdır:İlgili değişkenin değişim aralığıdır. Q3 - Q1 IQR'dır. Aslında robust değılım ölçüsüdür. Yani Q1 ve Q3 değerleri arasında zaten diğer değerlerin çoğu dağılıyordur. Bu değerler arasındakiler dikkate alınıyor. Kabul edinilebilirliği kanıtlayan temel noktanın çıkışı aslında brasıdır.

* İkisinin farkını aldık. Q1 = 20, Q3 = 40,aralarındaki fark ise 20 (yani IQR değerimiz = 20) olsun. Yukarıda görüldüğü gibi Q3'e IQR değerinin 1.5 katını ekliyoruz.(Resimde Q1 yazıyor ancak Q3 olacak. Yazım hatası olmuş...) Bu hesaplamalar sayesinde maksimum gidilebilecek upper sınırını belirliyoruz. Böylece biraz daha tolerans sağlamış oluyoruz. Bu üst sınırın üstünde olanlar da artık aykırı değerlerimiz olmuş oluyor.

* Alt sınırı da Q1 - 1.5 x IQR işlemini yaparak elde ediyoruz. Aynı şekilde bu alt sınırı geçemeyenler aykırı değerlerimiz oluyor.

* DİKKAT: Asıl nokta alt sınır ve üst sınırı belirleyebilmektir. Sonrasında aykırı değer değerlendirmesi yapabiliriz.

* DİKKAT: Eğer veri setimizin içersinde eksili değerler yoksa bu durumda (yani en küçük değerimiz 0 oluyor...) alt sınır çalışmıyor olacaktır, genelde üst sınıra göre çalışılıyor olur. Mesela yaş değişkeni için eksili bir değer olmayacaktır. Bu nedenle aşağıdan bir şeyler sili
