### PCA Grselletirme

```python
################################
# BONUS: PCA ile ok Boyutlu Veriyi 2 Boyutta Grselletirme
################################

################################
# Breast Cancer
################################

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

df = pd.read_csv("datasets/breast_cancer.csv")

df
```

**Amacmz** bu ok deikenli veriyi iki eksen zerinde grselletirmek istiyor olalm. Bunu ok fazla deiken olduundan dolay yapmak mmkn deil biliyoruz. yle bir ilem yapmalyz ki bunu iki boyutta grselletirebiliyor olalm.

```python
# Baml deikenimizi seelim:
y = df["diagnosis"]

# Bamsz deikenimizi seelim:
X = df.drop(["diagnosis", "id"], axis=1)

# Bu fonksiyona bamsz deikenleri ve baml deikeni verdiimizde nce bamsz deikenleri standartlatracak sonra da 
# pca hesab yapacak, deiken deerlerini dntrm olacak yani bileenleri karm olacak. Bu bileenleri bir dataframe'e
# dntrdkten sonra baml deiken ile yan yana concat ederek darya return edecektir.
def create_pca_df(X, y):
    X = StandardScaler().fit_transform(X)
    pca = PCA(n_components=2)
    pca_fit = pca.fit_transform(X)
    pca_df = pd.DataFrame(data=pca_fit, columns=['PC1', 'PC2'])
    final_df = pd.concat([pca_df, pd.DataFrame(y)], axis=1)
    return final_df

pca_df = create_pca_df(X, y)
pca_df

# Onlarca veri iki bileene indirgenmi oldu:
```

```python
# imdi bu iki bileeni grselletirmemiz lazmdr.

# Aadaki fonksiyon genel bir pca grseleltirme fonksiyonudur. Yani dataframe'i ve target' girdiimizde target'n esiz 
# snflarn bularak bunlar target adnda bir liste halinde tutacak, daha sonra her bir targeta gre seim ilemini yapp
# iki boyuta gre scatterplot oluturacak ve baml deikenleri grafiin zerine iaretliyor olacak. Her yerde bulunamayanlar
# serisinden bir fonksiyondur.
def plot_pca(dataframe, target):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xlabel('PC1', fontsize=15)
    ax.set_ylabel('PC2', fontsize=15)
    ax.set_title(f'{target.capitalize()} ', fontsize=20)

    targets = list(dataframe[target].unique())
    
    # iris problemi iin burada 3 tane snfmz vardr. Bu neden aadaki kodun colors = ['r', 'b'] olduu ekliyle yanl
    # bir kt elde ettik. Buradaki renk skalasn aadaki gibi dzenlersek doru kty elde edebiliriz. ki bileenli
    # bir problem iin colors = ['r', 'b'] dememiz yeterlidir, ancak 2'den fazla ise aadaki ekilde color list yapmalyz.
    # random.sample iin import random dedik en yukarda. Burada da import etmeyi tercih edebilirdik. Ancak burada import
    # edersek bu fonksiyonu arma sremiz uzar. Bir fonksiyonun iindeki importlar o fonksiyonun arlma sresini uzatr.
    colors = random.sample(['r', 'b', "g", "y"], len(targets))

    for t, color in zip(targets, colors):
        indices = dataframe[target] == t
        ax.scatter(dataframe.loc[indices, 'PC1'], dataframe.loc[indices, 'PC2'], c=color, s=50)
    ax.legend(targets)
    ax.grid()
    plt.show()

plot_pca(pca_df, "diagnosis")
```
