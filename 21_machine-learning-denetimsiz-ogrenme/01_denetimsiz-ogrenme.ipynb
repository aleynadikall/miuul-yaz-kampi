### Denetimsiz Öğrenme ( Unsupervised Learning )

**Denetimli Öğrenme:** Elimizdeki bazı featurelara karşılık bu featurelarda bir şeyler olduğunda ortaya çıktığı varsayılan bir çıktı yanıt bağımlı değişken olduğunda buna denetimli öğrenme problemleri deniliyordu. Eğer ilgilendiğimiz problemde bir bağımlı değişken/hedef değişken yoksa yani ilgili gözlem birimlerinde meydana gelen gözlem birimlerine karşılık orataya ne çıktığı bilgisi yoksa diğer ifadesiyle bir label/etiket yoksa bu durumda bunlar da gözetimsiz öğrenme yöntemleriydi.

Soldaki gözetimli sağdaki gözetimsiz öğrenmeyle ilgiligir.

   - K-Ortalamalar (K-Means)
   - Hiyerarşik Kümeleme Analizi (Hierarchical Cluster Analysis)
   - Temel Bileşen Analizi (Principal Component Analysis)

### Optimum Küme Sayısı Belirleme

```python
################################
# Optimum Küme Sayısının Belirlenmesi
################################

# Boş bir k-means nesnesi oluşturuyoruz:
kmeans = KMeans()

# Boş bir ssd listesi oluşturuyoruz:
ssd = []

# 1'den 30'a kadar K'lar oluşturuyoruz:
K = range(1, 30)

for k in K:
    kmeans = KMeans(n_clusters=k).fit(df)
    ssd.append(kmeans.inertia_)
```

### Hiyerarşik Kümeleme Analizi ( Hierarchical Cluster Analysis )

Amaç gözlemleri birbrilerine olan benzerliklerine göre alt kümelere ayırmaktır. Burada da yine kümelere ayırma işlemi vardır. Ancak hiyerarşik bir şekilde gerçekleşmektedir. Birleştirici ve bölümleyici olarak iki temel başlığımız vardır.



**Birleştirici kümelemede** çalışmanın başında bütün gözlem birimleri tek başına bir küme gibi düşünülüp yukarı doğru iki ya da üç gözlem birimi bir araya gelerek bir küme oluşturur. Ondan sonra bu iki ya da üç gözlemden oluşan iki küme bir araya gelerek başka bir küme oluşturur şeklinde yukarıya doğru birleştirerek işlemler yapılır.

**Bölümleyici kümelemede** ise tüm gözlemler bir arada olduğunda örneğin en yukarıdaki kısım tek kümedir. Sonrasında bu ayrılarak iki küme elde edilir. Ondan sonra ondan sonra vs şeklinde bölünerek bölümleyici şekilde aşağıya doğru tüm veri seti her bir gözlem tek başına kalacak şekilde aşağıya doğru oluşturulur.

**Peki bunun K-Means'den ne farkı vardır, ne alır ne götürür, nasıl yaklaşmak lazım?**

K-Means'te küme oluşturma sürecine dışarıdan müdahale edemiyorduk. Dolayısıyla bir gözlemleme imkanımız/şansımız yoktu. Ancak burada mesela şöyle bir şansımız var. Mesela diyebiliriz ki herhangi bir eyalet için bu eyalet burada tek başına kalmış o zaman bunu diğerleriyle bie kümeye sokalım, konuya buradan yaklaşalım diyebiliriz. Ya da istediğimiz sayıda küme seçebiliriz. 

### Hiyerarşik Kümeleme ( Hierarchical Cluster )

```python
################################
# Hierarchical Clustering
################################

df = pd.read_csv("datasets/USArrests.csv", index_col=0)

# Veri setini standartlaştıralım:
sc = MinMaxScaler((0, 1))
df = sc.fit_transform(df)

hc_average = linkage(df, "average")

# Hiyerarşik kümeleme yöntemlerinin temek noktası, temel yapıtaşı dendogram adını verdiğimiz o kümeleme yapısını gösteren 
# şemadır.
plt.figure(figsize=(10, 5))

# Title bilgisi
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_average,
           # İismlendirme boyutu
           leaf_font_size=10)
plt.show()
```

**Linkage**


```python
# Daha az sayıda küme görmek istersek truncate_mode="lastp" kullanmalıyız:
plt.figure(figsize=(7, 5))
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_average,
           truncate_mode="lastp",
           p=10,
           show_contracted=True,
           leaf_font_size=10)
plt.show()
```

Hiyerarşik kümeleme yönteminin avantajı teori yönteminde de ifade edildiği gibi bize gözlem birimlerine genelden bakma şansı tanımasıdır ve buna göre çeşitli karar noktalarına dokunabilme şansı tanımasıdır.

### Temel Bileşen Analizi ( Principal Component Analysis )



Diyelim ki elimizde 100 tane değişkenden oluşan bir veri olsun. Küçük bir bilgi kaybını göze alıp bu 100 değişkeni örneğin 3 değişkene indirme çabasıdır. 100 tane değişken yerine 3 tane değişken ile veri seti içerisindeki değişkenliği/varyansı/bilgiyi temsil ediyor oluruz. Bunu bedeli ufak miktarda göz ardı edilebilecek miktarda bilgiden fedakarlık yapmaktır. İyi de neden böyle bir şeyler uğraşıyoruz, boyut indirgeme gibi bir kaygımız var gibi bir soru olabilir. Bunun bazı yanıtları vardır. Örneğin doğrusal regresyon yöntemlerinde çoklu doğrusal bağlantı problemlerinden kurtulmak istiyor olabiliriz, örneğin bir yüz tanıma probleminde resimlere filtre yapma ihtiyacı hissediyor olabiliriz. Tabii ki bu çeşitli sosyal medya uygulamalarındaki filtrelemeler değildir. Bir gürültü azaltma aracı olarak kullanabiliriz ya da buna benzer sebeplerle bir şekilde boyut indirgeme ihtiyacımız olduğunda bu yöntemi kullanırız.

**DİKKAT :** Eğer bir mülakatta bir şekilde principal bileşen analizi yöntemi nedir ne yapar diye sorulursa temel bileşen analizi bir boyut indirgeme yaklaşımıdır. Bu şu anlama gelir, veri setinin boyutunun küçük miktar bir bilgi kaybını göze alarak indirgeme işlemidir dememiz gerekir.

Aşağıdaki görselde soldaki veriye PCA uyguladığımızda 3 boyuttan 2 boyuta düşerek sağdaki sonuç elde edilir.



Görüntü işleme problemlerinde regresyon problemlerinde bazı varsayım problemlerinde yani bir şekilde çok boyutluluk söz konusu bir şekilde boyut indirgenmek isteniyorsa PCA kullanılabilir.

**DİKKAT :** PCA yöntemi birçok farklı konu birçok farklı alanda kullanılabilen temel bir tekniktir. Dolayısıyla sadece direkt bir şey yapmak için kullanılmaz. Örneğin çok boyutlu veriyi görselleştirmek için de kullanılır. Mesela elimizde 50 tane değişken var, 50 değişken demek 50 boyut demektir. 50 değişkeni nasıl görselleştireceğiz? Bir şey lazım bize bunu 2 boyuta indirgeyebilecek. PCA bunun için kullanılabilecek bir tekniktir. Veri görselleştirme problemimizi PCA ile çözmüş olduk.

**DİKKAT :** Temel bileşen analizi veri setini bağımsız değişkenlerin doğrusal kombinasyonları ile ifade edilen bileşenlere indirger. Dolayısıyla bu bileşenler arasında korelasyon yoktur. Bu bileşenler korelasyonsuzdur. Mesela birbirleri arasında korelasyonu olan elimizdeki üç değişkeni PCA ile 2 boyuta indirgediğimizde aralarındaki korelasyon ortadan kalkar. Temel matematiğine çok girmiyor olacağız. 

Peki teknik olarak detaylarda bu bileşenleri nasıl çıkarırız sorusunun yanıtı ise değişken gruplarının varyanslarını ifade eden özdeğerler ve veri setindeki değişkenleri gruplandırır. Yani gruplar arasında en fazla varyansa sahip olan gruplar en önemli gruplardır ki işte bunlar asal bileşenlerdir. Yani varyansa dayalı, özdeğerlere dayalı bir gruplama yapılmaktadır.

### Temel Bileşen Regresyon Modeli ( PCR )

```python
################################
# BONUS: Principal Component Regression
################################

# Diyelim ki Hitters veri seti doğrusal bir model ile modellenmek istiyor ve (varsayım bunlar) değişkenler arası çoklu doğrusal 
# bağlantı problemi var. Bu doğrusal regresyon modelelrinde sağlanması gereken varsayımlardandır. Değişkenler arasında yüksek
# korelasyon olduğunda bu çeşitli problemlere sebep olur. Bunu istemiyor oluruz. Diyelim ki böyle bir problemimiz var.
# DİKKAT : 1.si bu amacımızı yerine getireceğiz. 2.si yukarıda çıkarmış olduğumuz bileşenlerin neye karşılık geldiğini amlamaya
# çalışacağız birlikte.

df = pd.read_csv("datasets/Hitters.csv")
df.shape
```

```python
len(pca_fit)
```

```python
# Numerik sütun sayısına bakalım:
num_cols = [col for col in df.columns if df[col].dtypes != "O" and "Salary" not in col]
len(num_cols)
```

```python
# num_cols dışında kalan değişkenleri getiriyoruz:
others = [col for col in df.columns if col not in num_cols]

# Bileşenleri DataFrame'e çeviriyoruz:
pd.DataFrame(pca_fit, columns=["PC1","PC2","PC3"]).head()

# Daha öncesinde 16 tane sayısal değişkenimşz vardı, biz bunları 3 taneye indirgedik. Bunun yanında bir de veri setindeki 
# bağımlı değişken var, diğer kategorik değişkenler var. Ne yapacağız? Şöyle bir amacımız var: PCR denilen bşr makine öğrenmesi
# yöntemi vardır. Temel bileşen regresyonu yöntemi vardır. Bu yöntem şöyle çalışıyor. Önce temel bileşen yöntemi kullanılıp 
# değişkenlerin boyutu indirgeniyor. Daha sonrasında bu bileşenlerin üzerine bir regresyon modeli kuruluuyor. Bunun direkt
# pythonda uygulandığı bir metod yoktur. O zaman kendimiz uyarlayacağız bunu.
```

```python
df[others].head()
```

```python
# Yukarıda oluşturduğumuz veri setini bir araya getirelim:
final_df = pd.concat([pd.DataFrame(pca_fit, columns=["PC1","PC2","PC3"]),
                      df[others]], axis=1)
final_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

# Model kurmak istiyoruz ancak bir problemimiz var, buradaki kategorik değişkenler... Kategorik değişkenlerin hepsinin sınıf 
# sayısı iki olduğundan dolayı bunları dilersek direkt label_encoder'dan geçirebiliriz ya da get_dummies'i kullanabiliriz ya da
# kendi tanımlamış olabileceğiz başka bir one_hot_encoder'ı (drop_first=True olarak) kullanabiliriz. 
def label_encoder(dataframe, binary_col):
    labelencoder = LabelEncoder()
    dataframe[binary_col] = labelencoder.fit_transform(dataframe[binary_col])
    return dataframe

for col in ["NewLeague", "Division", "League"]:
    label_encoder(final_df, col)
    
final_df.head()
```

```python
# NaN'lerden kurtulmak için:
final_df.dropna(inplace=True)
```

```python
final_df.head()
```

```python
y = final_df["Salary"]
X = final_df.drop(["Salary"], axis=1)

# Bunu daha sonra kullanmak gibi bir derdiğimiz olmadığından fit etme işlemini yapmıyoruz. Direkt hatalarına bakacağız:
lm = LinearRegression()
rmse = np.mean(np.sqrt(-cross_val_score(lm, X, y, cv=5, scoring="neg_mean_squared_error")))
rmse
```

```python
# rmse değerimiz 345 çıktı. Doğru yolda mıyız, yanlış yolda mıyız, buna nasıl karar veriyorduk? Bunun için y bağımlı 
# değişkeninin ortalamasını alarak değerlendiriyorduk. 545 çıktı. İyi değildir ancak çok kötü de değildir. Çünkü zaten radikal 
# bir işlem yaptık. Temel bileşen analizi diye birşey öğrendik. 16 tane değişkeni bir miktar bilgi kaybını göze alarak birden
# indirgedik. Böyle radikal bir hamleye rağmen başarımız oldukça kayda değer, oldukça doğru yolda olduğumuzu düdündürebilecek
# seviyededir.

y.mean()
```

```python
# DecisionTreeClassifier ile de bir değerlendirelim:
cart = DecisionTreeRegressor()
rmse = np.mean(np.sqrt(-cross_val_score(cart, X, y, cv=5, scoring="neg_mean_squared_error")))

rmse
# Yani regresyon modelinden daha kötü gibi gözüküyor. Şimdi buna bir hiperparametre optimizasyonu yapmazsak bu ağlamaz mı?
```

```python
# Hiperparametre optimizasyonu yapalım:
cart_params = {'max_depth': range(1, 11),
               "min_samples_split": range(2, 20)}

# GridSearchCV
cart_best_grid = GridSearchCV(cart,
                              cart_params,
                              cv=5,
                              n_jobs=-1,
                              verbose=True).fit(X, y)

cart_final = DecisionTreeRegressor(**cart_best_grid.best_params_, random_state=17).fit(X, y)

# Son modelimizi kurduk, 10 katlı çapraz doğrulama ile hatalarımızı bir değerlendirmek istediğimizi düşünelim:
```

### PCA Görselleştirme

```python
################################
# BONUS: PCA ile Çok Boyutlu Veriyi 2 Boyutta Görselleştirme
################################

################################
# Breast Cancer
################################

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

df = pd.read_csv("datasets/breast_cancer.csv")

df
```

**Amacımız** bu çok değişkenli veriyi iki eksen üzerinde görselleştirmek istiyor olalım. Bunu çok fazla değişken olduğundan dolayı yapmak mümkün değil biliyoruz. Öyle bir işlem yapmalıyız ki bunu iki boyutta görselleştirebiliyor olalım.

```python
# Bağımlı değişkenimizi seçelim:
y = df["diagnosis"]

# Bağımszı değişkenimizi seçelim:
X = df.drop(["diagnosis", "id"], axis=1)

# Bu fonksiyona bağımszı değişkenleri ve bağımlı değişkeni verdiğimizde önce bağımsız değişkenleri standartlaştıracak sonra da 
# pca hesabı yapacak, değişken değerlerini dönüştürmüş olacak yani bileşenleri çıkarmış olacak. Bu bileşenleri bir dataframe'e
# dönüştürdükten sonra bağımlı değişken ile yan yana concat ederek dışarıya return edecektir.
def create_pca_df(X, y):
    X = StandardScaler().fit_transform(X)
    pca = PCA(n_components=2)
    pca_fit = pca.fit_transform(X)
    pca_df = pd.DataFrame(data=pca_fit, columns=['PC1', 'PC2'])
    final_df = pd.concat([pca_df, pd.DataFrame(y)], axis=1)
    return final_df

pca_df = create_pca_df(X, y)
pca_df

# Onlarca veri iki bileşene indirgenmiş oldu:
```

```python
# Şimdi bu iki bileşeni görselleştirmemiz lazımdır.

# Aşağıdaki fonksiyon genel bir pca görselelştirme fonksiyonudur. Yani dataframe'i ve target'ı girdiğimizde target'ın eşsiz 
# sınıflarını bularak bunları target adında bir liste halinde tutacak, daha sonra her bir targeta göre seçim işlemini yapıp
# iki boyuta göre scatterplot oluşturacak ve bağımlı değişkenleri grafiğin üzerine işaretliyor olacak. Her yerde bulunamayanlar
# serisinden bir fonksiyondur.
def plot_pca(dataframe, target):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xlabel('PC1', fontsize=15)
    ax.set_ylabel('PC2', fontsize=15)
    ax.set_title(f'{target.capitalize()} ', fontsize=20)

    targets = list(dataframe[target].unique())
    
    # iris problemi için burada 3 tane sınıfımız vardır. Bu neden aşağıdaki kodun colors = ['r', 'b'] olduğu şekliyle yanlış
    # bir çıktı elde ettik. Buradaki renk skalasını aşağıdaki gibi düzenlersek doğru çıktıyı elde edebiliriz. İki bileşenli
    # bir problem için colors = ['r', 'b'] dememiz yeterlidir, ancak 2'den fazla ise aşağıdaki şekilde color list yapmalıyız.
    # random.sample için import random dedik en yukarıda. Burada da import etmeyi tercih edebilirdik. Ancak burada import
    # edersek bu fonksiyonu çağırma süremiz uzar. Bir fonksiyonun içindeki importlar o fonksiyonun çağırılma süresini uzatır.
    colors = random.sample(['r', 'b', "g", "y"], len(targets))

    for t, color in zip(targets, colors):
        indices = dataframe[target] == t
        ax.scatter(dataframe.loc[indices, 'PC1'], dataframe.loc[indices, 'PC2'], c=color, s=50)
    ax.legend(targets)
    ax.grid()
    plt.show()

plot_pca(pca_df, "diagnosis")
```
