###### Regresyon Problemlerinde Başarı Değerlendirme Metrikleri

* 
* sales değişkeni ve diğer değişkenler arasındaki ilişkinin modellendiğini ve neticesinde bir makine öğrenmesi yöntemi aracılığıyla bu satışların tahmin edilebilir bir hale geldiğini varsayalım, bu tahminler de aşağıdaki gibi olsun.

* 
* Hatalı tahminler oluştuğunu gözlemliyoruz. Bunlara artık, hata, error denir. 

* 
* Bu formül çok çok iyi bilmemiz gereken makine öğrenmesi temellerinde kendine sağlam bir yer edinmiş hem bir başarı değerlendirme ölçüsü, hem de optimizasyon yöntemlerini kullandığımızda daha iyi bir model kurmak üzere kendisini optimize etmeye çalıştığımız bir fonksiyondur. Bunu adımız soyadımız gibi bilmemiz gerekir. Kurmuş olduğumuz model neticesindeki yi (gerçek değerler) ile model aracılığı ile elde edilen tahmin ettiğimiz değerler arasındaki farkları değerlendirme imkanı sağlar.

* DİKKAT : Buradaki yi ifadesi veri setindeki gerçek satış değerlerini ifade etmektedir. Şapkalı yi ifadesi ise model aracılığıyla bizim tahmin ettiğimiz değerlerdir. Dolayısıyla burada gerçek değerler ile tahmin edilen değerler arasındaki farkın karelerinin toplamı alınmış, sonrasında da ortalaması alınmıştır. Kısacası yukarıdaki ifade modelimizin başarısını ölçeceğimiz ortalama hata metriğidir. 

* 
* Bu da kullanılabilir, farkı min square error'da hataların farkını aldıktan sonra bir kare işlemi varken root min square error'da bu kare ifadesini geri dönüştürmek gibi görülecek karekök alma işlemi vardır.

* 
* Bir diğeri de mutlak ortalama hatadır.
