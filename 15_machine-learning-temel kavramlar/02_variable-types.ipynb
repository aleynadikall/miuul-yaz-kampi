### Değişken Türleri (Variable Types)
* Sayısal Değişkenler : Yaş, metrekare, fiyat değişkeni. Kesikli (yaşın tam sayı olması) veya sürekli (küsüratlı, float değerler) olabilirler.
* Kategorik Değişkenler(Nominal (sınıflar arasında farklılık olmaması), Ordinal (mesela eğitim durumu)) : Kadın-Erkek (Nominal), Hayatta Kaldı-Hayatta Kalamadı, Futbol Takımları (Nominal). Yani bir sınıfı kategoriyi ifade eden değişkenlerdir.
* Bağımlı Değişken (target, dependennt, output, response) : Makine öğrenmesinde supervised (denetimli, gözetimli) problemlerinde karşımıza çıkar. Hastalık tahminiyle ilgileniyor olduğumuzda hasta olup olmama durumunu ifade eden değişken bağımlı değişkendir. Ev, araç fiyat tahminiyle ilgileniyor olduğumuzdda ev ve araçların fiyatları bağımlı değişkenlerdir.
* Bağımsız Değişken (feature, independent, input, column, predictor, explanatory) : İlgilendiğimiz problemdeki bağımlı değişkeni oluşturduğunu, etki ettiğini varsaydığımız diğer değişkenlerdir.

* 
* Yukarıdaki tablodaki bağımlı değişken satış (sales) değişkenidir. Bu problem bir regresyon problemidir çünkü bağımlı değişkenimiz sayısaldır.

* 
* Churn : Müşteri Terki demektir. Buradaki bağımlı değişkenimiz churn'dür. Total_Purchase da göreceli olarak eğer bir soru olarak aklımıza gelirse odaklanıldığında bir bağımlı değişken gibi düşünülebilir ama bu veri setinde bu problem elimize geldiğinde bağımlı değişkeni biliyor oluruz. Chur değişkeni sayısal olarak encode edilmiş iki sınıflı kategorik değişkendir. Bu bir sınıflandırma problemidir çünkü bağımlı değişken kategorik değişkendir.
