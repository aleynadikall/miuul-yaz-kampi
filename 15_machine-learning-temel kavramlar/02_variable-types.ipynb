### Deiken Trleri (Variable Types)
* Saysal Deikenler : Ya, metrekare, fiyat deikeni. Kesikli (yan tam say olmas) veya srekli (ksratl, float deerler) olabilirler.
* Kategorik Deikenler(Nominal (snflar arasnda farkllk olmamas), Ordinal (mesela eitim durumu)) : Kadn-Erkek (Nominal), Hayatta Kald-Hayatta Kalamad, Futbol Takmlar (Nominal). Yani bir snf kategoriyi ifade eden deikenlerdir.
* Baml Deiken (target, dependennt, output, response) : Makine renmesinde supervised (denetimli, gzetimli) problemlerinde karmza kar. Hastalk tahminiyle ilgileniyor olduumuzda hasta olup olmama durumunu ifade eden deiken baml deikendir. Ev, ara fiyat tahminiyle ilgileniyor olduumuzdda ev ve aralarn fiyatlar baml deikenlerdir.
* Bamsz Deiken (feature, independent, input, column, predictor, explanatory) : lgilendiimiz problemdeki baml deikeni oluturduunu, etki ettiini varsaydmz dier deikenlerdir.
