### Yanlılık - Varyans Değiş Tokuşu (Bias - Variance Tradeoff)

* Bu bölümde aşırı öğrenme konusunu değerlendiriyor olacağız. 
* Bir model geliştirdiğimizi düşünelim ve bu model neticesinde bazı hata metriklerine eriştiğimizi düşünelim. Yine genel resimde kullanabileceğimiz onlarca, yüzlerce makine öğrenmesi yöntemi vardır. Bunlardan herhangi bir tanesini kullanıp bunun neticesinde başarımızla ilgili bir değerlendirme yapıyor olduğumuzu düşünelim. Aşırı öğrenme uygulamalı makine öğrenmesi kapsamında karşılaşılan en ciddi problemlerden birisidir. 

###### Overfitting (Aşırı öğrenme) Nedir?

* 

* Çok büyük ihtimalle bir mülakatta karşımıza çıkabilir. Modelin veriyi öğrenmesidir. Model veriyi değil veri içindeki yapıyı, örüntüyü öğrenmelidir. Aşırı öğrenme verinin ezberlenmesidir. Bu durumun karşılığı yüksek varyanstır.

###### Underfitting Nedir?

* Modelin veriyi öğrenememesidir. Yukarıdaki grafiklerde kırmızılar tahmin fonksiyonu maviler gerçek değerlerdir. Yüksek yanlılık vardır. Bazı gözlemlere daha yanlıdır.

* DİKKAT : Bir çocuk olduğunu düşünelim. Bu çocuğun bir sınavı var ve sınavdan önce öğretmen bu çocuğa falanca sorulara çalış da gel demiş olsun. Çocuğun o soruları birebir ezberlediğini düşünelim. Yani hoca aslında oradaki soruların yapısını öğrenmesi, mantığını anlaması için o soruları vermişti fakat öğrenci öğrenmek yerine ezberledi. Ertesi gün bu çocuk sınava girdiğinde önüne sorular geldiğinde o sorular sorulan sorunun mantığını taşısa da öğrenci çözemeyecektir. Bu duruma ezberleme denir. 

* Bizim aradığımız yanlılığı ve varyansı düşük doğru modeldir. Doğru modelin temsili de veri setindeki ilişkinin, yapının, örüntünün temsil edilmesidir.

* DİKKAT : Mülakatta karşımıza aşırı öğrenmeye düştüğünüzü nasıl anlarsınız sorusu gelebilir. Çok kritik bir sorudur. Pratikte de yakalanması bir iş olan bir sorudur. 

###### Aşırı Öğrenme Nasıl Tespit Edilir?

* 

* Aşırı öğrenme sorunu eğitim seti ve test setinin model karmaşıklığı ve tahmin hatası çerçevesinde birlikte değerlendirilmesi ile tespit edilebilir. Eğitim seti ve test setindeki hata değişimleri incelenir ve bu iki hatanın birbirinden ayrılmaya çatallannmaya başladığı nokta itibari ile aşırı öğrenme başlamıştır denir. Dolayısıyla aşırı öğrenmenin başladığı noktayı yakalayabiliyorsak aşırı öğrenmenin nasıl önüne geçebiliriz sorusunun yanıtlarından birisini de bulmuş oluruz. 

###### Model Karmaşıklığı Nedir?
* Bu sorunun cevabı doğrusal modellere göre farklı, ağaç yöntemlere göre farklı, sinir ağlarına göre farklı olacaktır. Örneğin doğrusal yöntemlerde modele üstel terimler eklemek yani modeli hassaslaştırmak, daha detaylı tahminler yapabilir hale getirmek olacaktır ki modeli karmaşıklaştırmak demek olacaktır. Doğrusal formülasyon olduğunu düşünelim (y = b + wx). Buraya x'in karesini/küpünü eklediğimizde bu durumda bu doğrusal yöntemler için model karmaşıklığıdır. Ağaç yöntemlerde ise dallanma sayısıdır. Dallanma sayısının artması da ağaç yöntemlerde ilgili modelin karmaşıklaşmasına sebep olacaktır. Karmaşıklaştırmak kötü değildir. Bir süreye kadar hatalarımızı düşürüyor olacaktır. Ancak bir noktadan sonra train setini çok iyi öğrendiğinden dolayı fazla karmaşıklaştırmamak gerekir. Lightgbm'de ise iterasyon sayısı arttıkça model karmaşıklaşacaktır (Gradient descent algoritması üzerinde optimize edilerek oluşturulan ağaçlar...). Yapay sinir ağları kullanıldığını düşünelim, burada da katman sayısı, hücre sayısı, iterasyon sayısı arttırıldığında model karmaşıklığı artacaktır.

* Yukarıdaki grafiğe baktığımızda test seti eğitim setinin peşinden gidiyor, gayet güzel. Model karmaşıklığının artıyor olması daha iyi model yapıyoruz demektir. Az öğrenmeden çok öğrenmeye gidiyoruz. Eğitim seti bir miktar hatayı düşürmeye devam ediyor olacaktır. Fakat bir noktadan sonta test seti şunu diyecektir. Ya ben senle buraya kadar geldim, iyisin hoşsun sen de hata düştükçe bende de düştü çok teşekkür ederim sizin model iyiymiş her şey yolunda ama bir noktadan sonra bu model sana takıntılı hale gelmeye başladı sanırım diyecektir test seti. Bu noktadan sonra da test setinin hatası artmaya başlayacaktır. Model eğitim setine takıntılı bir hale geleceğinden ilişki öğrenmekten bu veriyi öğrenmeye dönüşeceğinden dolayı bu veriyi çok iyi öğreniyor olacaktır fakat test setine gittiğinde test hatası artıyor olacaktır. Dolayısıyla şu anda olası bir mülakatta gelme ihtimali çok yüksek olan iki üç soruya cevap bulmuş olduk.

###### Aşırı Öğrenme Sorunu Nasıl Çözülür?

* Aşırı öğrenme sorunu veri setinin boyutunun arttırılması, feature selection yapılabilir ama en nihayetinde değişkenler ve gözlemlerde ya da çeşitli diğer bazı faktörlerde değişiklik olmayacağını varsayarsak eğitim setiyle test setinin hatalarının birbirinden ayrışmaya başladığı nokta optimum nokta olarak seçilir, bu noktada durulduğunda aşırı öğrenmenin önüne geçilmiş olunur.
