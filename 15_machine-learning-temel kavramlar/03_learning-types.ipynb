###### Pekiştirmeli Öğrenme
* Boş bir oda olduğunu düşünelim. Bunun içerisinde bir robot olduğunu ve bu robotun bu boş odadan kapıyı kullanarak çıkma görevi olduğunu farzedelim. Robotun yapabileceği olası hamleler bir adım ileri, sağa, sola, geri gitmek, zıplamak, elini, kolunu kaldırmaktır. Bu robot deneme yanılma yoluyla bu odadan çıkmaya çalışıyor olsun ve başarısız olduğu her girişimiyle de cezalandırılıyor. Bu şekilde deneme yanılma yoluyla yanlış yaptığı şeylerden pekiştire pekiştire o kapıdan çıkabiliyor olacaktır.

###### Denetimli, Gözetimli Öğrenme

* 
* Eğer veri setimizde labellarımız yer alıyorsa bu durumda veri denetimli öğrenme problemine sahiptir. Label ifadesi daha çok sınıflandırma problemlerinde görülüyor olur. Ancak regresyonda da gözlemlediğimiz durumlar vardır. Bağımlı ve bağımsız değişkenler arasındaki ilişki öğreniliyor olur.
