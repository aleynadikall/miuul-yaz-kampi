###### Cross Validation

* Bu yntemde versi setini 4 para eitim seti, 1 para test seti olacak ekilde 5 paraya blyoruz. Bu paralar kendi arasnda model kurulup test edildikten sonra srekli deiiyor. Yeni 4 paralk train setiyle modeli kuruyoruz ve yeni 1 paralk test setiyle modeli test ediyoruz. Bu srekli tekrar ediyor. Sonularn ortalamasn alarak bize bir sonu retir. Gzlem says az olduunda kullanmak daha gzel olur.
