### Model Doğrulama (Model Validation) Yöntemleri

* Önceki bölümlerde elde ettiğimiz hataları (MSE, Accuracy vb.) daha doğru değerlendirme ve doğrulamaya çalışma çabasına model doğrulama diyebiliriz. 

###### Holdout Yöntemi (Sınama Seti Yöntemi)

* Eski dünyada makine öğrenmesi yöntemleri kullanıldıkları veri seti üzerinde başarı açısından değerlendirilirdi. Yani bir modelin başarısı o modelin eğitmek için kullanılan veri seti ile değerlendirilirdi. Dolayısıyla bir veri seti üzerinde hem model kurulurdu, hem de başarısı değerlendirilirdi.

* 

* Model doğrulama yöntemlerinde amaç veri seti eğitim ve test seti olarak ikiye ayrılır, eğitim seti üzerinde train işlemi gerçekleştirilir böylece model öğrenmiş olur, daha sonra o modele test setinden sorular sorulur, kendini burada test eder.

###### K-Katlı Çapraz Doğrulama (K Fold Cross Validation)

* Diyelim ki orjinal veri setimizde 100 tane gözlem birimi olsun. 0 ve 1'lerden oluşan bir sınıflandırma problemi olsun. Diyelim ki veri setini %80'ini eğitim seti, %20'si test seti olacak şekilde 2 parçaya böldük. Ratgele gerçekleştirilen bu bölme işleminde diyelimki 20'lik parçadaki tüm gözlem birimlerinin bağımlı değişken değerleri 1'dir, ya da bağımsız değişken değerleri verinin çok genel olmayan yapısını temsil ediyor olsun. Dolayısıyla test sonucunda 20'lik parçadan elde edilen sonuç bir miktar yanıltıcı olacaktır. 
* DİKKAT : Makine öğrenmesi dünyasına girdiğimizde ortam biraz daha fludur. Tek bir doğru yoktur. Buradaki senaryo gözlem sayısının az olması olduğundan ortaya çıkabilecek test train kısımları çok da sağlıklı olmayabilir. Bu bölme işlemi bize problem yaratabilir. 

* 

###### Cross Validation

* Bu yöntemde versi setini 4 parça eğitim seti, 1 parça test seti olacak şekilde 5 parçaya bölüyoruz. Bu parçalar kendi arasında model kurulup test edildikten sonra sürekli değişiyor. Yeni 4 parçalık train setiyle modeli kuruyoruz ve yeni 1 parçalık test setiyle modeli test ediyoruz. Bu sürekli tekrar ediyor. Sonuçların ortalamasını alarak bize bir sonuç üretir. Gözlem sayısı az olduğunda kullanmak daha güzel olur.

###### K Fold

* Bu yöntemde de veri seti önce eğitim seti ve test seti diye ikiye ayrılır (Holdout Yöntemi). Sonrasında da eğitim seti 5 parçaya bölünerek yukarıda anlatılan çapraz doğrulama eğitim seti üzerine uygulanır. Sanki tüm veri eğitim setiymiş gibi davranılır. En son da modelin hiç görmediği başta ayırdığımız test setine gider ve tekrar test edilir. Veri seti bol ise bu durumda K Fold kullanılabilir.
