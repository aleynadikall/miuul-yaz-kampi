###### Özellikleri verilen kişilerin hayatta kalıp kalamayacağını tahmin edebilir misiniz?

* 

* Yukarıda bir titanic faciası veri seti vardır. Bu veri setindeki insanlar çeşitli özelliklere sahip kişilerin bazısı hayatta kalıyor bazısı kalamıyor. Hayatta kalma durumunu 1, hayatta kalamama durumunu 0 ifade ediyor. Survived değişkeni bu çalışmadaki target, bağımlı değişkendir. Ayrıca 0 ve 1 gibi kategorik olarak encode edilmiş kategorik bir değişkendir. Yani bu bir sınıflandırma problemidir. Yukarıdaki problemimizde bağımlı değişkenimiz sayısal bir değişkendi, bu problemde ise kategorik değişkendir. Target değişkenimiz sınıfları ifade eden bir değişken olduğunda (kategorik değişken) bu bir sınıflandırma problemidir. Survived değişkenine ek olarak da 3 tane bağımsız değişkenimiz var.

###### 45 yaşında, 3.sınıf ve erkek olan bir yolcu hayatta kalabilir mi?
* Tabloda 45 yaşı göremiyoruz ama yakınındaki yaşlara gidebiliriz. Görüyoruz ki cinsiyeti erkek olup 45 yaşa en yakın olan 35 yaş var. Bu yolcunun yolculuk sınıfı ise 1 olarak kaydedilmiş ancak biz yolculuk sınıfı 3 olan bir erkek yolcu istiyoruz, buradan da bir yere gidemedik. O zaman tek tek inceleyelim. 
* 1.satır : .sınıf erkek yolcu ve yaşı 22 ancak hayatta kalamamış.
* 2., 3., 4.satırdakilerin cinsiyeti kadın ve hepsi hayatta kalmıştır. Demek ki bu gemide kadın olmanın hayatta kalmada büyük bir önemi var.
* 5.satır : Erkek, 1.sınıf yolcu ve 35 yaşında hayatta kalmış.
* 6.satır : Bir başka erkek ve yine 1.sınıf yolcu, 25 yaşında bu da hayatta kalmış. 
* 7.satır : Yaşı 25 olmasına rağmen 2.sınıf yolcu erkek yolcu hayatta kalamamış. Sanki kadın olmakla beraber yolculuk sınıfı da bir kişinin hayatta kalması durumuna etki edebilir gibi gözüküyor.
* 8.satır : 3.sınıf bir yolcu, erkek ve yaşı 15. Yaşı çok küçük olmasına rağmen hayatta kalamamış.
* Bu durumda bu kişi erkek ve 3.sınıf yolcu olduğu için hayatta kalamaz diyoruz.

###### 15 yaşında, 1.sınıf, erkek bir yolcu hayatta kalabilir mi?
* 1.sınıf dediği için bir avantaja sahip gibi düşünüyoruz. Çünkü 1.sınıf olarak gözlemlediğimiz 3 gözlem var ve bunların hepsi hayatta kalmıştır. Yaş bilgisine baktığımızda belirli bir örüntü yakalayamadığımız için kesin bir şey diyemiyoruz. Ancak yaşı küçük olduğu ve 1.sınıf yolcu olduğu için hayatta kalabilir diyoruz. 

###### 36 yaşında, 1.sınıf, kadın yolcu hayatta kalabilir mi?
* Elimizdeki veriye göre tereddüt etmiyoruz. Hayatta kalır diyoruz. 

* 3 sorunun da cevabını verirken farklı kesinliklerde konuştuk. Çünkü yolcuların hayatta kalma olasılıklarını bilmiyor olma kaygımız var. 

###### 40 yaşında, 3.sınıf, kadın yolcu hayatta kalabilir mi?
* Kadın olduğu için pozitif bir önyargımız var. Tabloya dikkat ettiğimizde 1.sınıfın ve kadın cinsiyetinin bu veri setindeki eziciliğini biliyoruz. Ancak 3.sınıf olup hayatta kalan kadın yolcunun yaşı tablomuzda 12. Yaşı bu kadar yük olup (40) hayatta kalan 3.sınıf bir yolcuyla ilgili kesin bir şey diyemiyoruz. Kadın olması durumundan dolayı pozitife geçiyoruz ancak yaştan dolayı tereddüte düşüyoruz. Yine de buna hayatta kalabilir diyoruz.
