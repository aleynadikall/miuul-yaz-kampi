### Özellik Çıkarımı (Feature Extraction)
* Veri bilimi ve makine öğrenmesi süreçlerinde oldukça önemli olan feature engineering ile eş anlamlarda da kullanılabilen değerli bir konudur.
* Özellik çıkarımı ham veriden değişken(özellik-feature) çıkarmak demektir.
* Ham veriden değişken üretmek iki kapsamda düşünülebilir.
* Birincisi yapısal verilerden değişkenler türetmek.
* İkincisi yapısal olmayan verilerden değişken türetmek.
* Yapısal verilerden değişken türetmek dediğimiz elimizde zaten belirli bir yapıda olan mevcut değişkenler üzerinden değişken türetmek, diğeri ise yapısal olmayan yani görüntü gibi ses gibi yazı gibi verilerden özellik türetmektir. İyi de neden böyle özellikler üretmek istiyoruz? Numerik veriler üzerinde çalışabiliyoruz, tabular data dediğimiz excel tarzındaki veriler üzerinde çalışabiliyoruz. Ama elimizde bir resim, yazı, ses olduğunda bazı öğrenme algoritmalarını kullanmak için bu yapısal olmayan şeyleri matematiksel olarak anlaşılabilecek şeylere yani numerik temsillere dönüştürmemiz gerekir. Bundan dolayı yapısal olmayanları da dönüştürmek ve bunlardan özellikçıkarmak işlemiyle ilgileniyor oluruz.

* Öncelikle yapısal olabilecek değişkenlerden özellikler çıkarma, değişkenler çıkarma durumlarını değerlendirelim:

* 
* Diyelim ki elimizdeki yukarıdaki gibi bir değişken var. Buradan türetilebilecek yeni değişkenler nelerdir sorusunun cevabını arıyoruz. Timestamp bir zaman değişkenidir. Bunu içerisinde yıl, ay, gün ve saat bilgisi vardır. Dolayısıyla bu normalde tek olan değişkenden yıl, ay, gün, saat değişkeni çıkabilir. Sadece bunlarla sınırlı değildir. 

* 

* Mesela buradaki günlerin haftanın hangi günlerine denk geldiğini de bulabiliriz ve gün isimlerini çıkarabiliriz. Buna benzer şekilde bu veri seti üzerinden türeyebilecek birçok farklı değişken de olabilir. İşte bu durum yapısal, varolan bazı değişkenler üzerinden yeni değişkenler tüüretmeye örnek olarak gösterilebilir.
* Şimdi bir başka soruya cevap verelim. Diyelim ki elimizde aşağıdaki gibi metinsel bir ifade var. Bu titanic veri setindeki Name değişkenindeki isimler. Buradan hangi yeni değişkenler türetilir, türetilebilir mi, türetilemez mi?


* DİKKAT: Mesela Mrs ve Mr ifadeleri, bu titlelar aslında bir bilgi barındırmaktadır. Normalde bu değişken tek başına kardinalliği yüksek seyrekliği yüksek bilgi taşımama durumu olan bir değişken fakat sanki buna yakınlamasına bakarsak aslında buradaki titlelardan, isimlerden hareketle sanki kişilerle ilgili artık ölçülebilir bir ya da birden fazla değişken türetebiliriz gibi geliyor. Bu titlelara göre kişilerin cinsiyeti ortaya çıkabilir. Yine bu titlelara göre kişilerin yaşlarına ilişkin bilgiler çıkabilir. 
* DİKKAT: Daha güzel bir senaryoo... Alttan 2.satırda Dr.ifadesi var. Belki doktorlar titanik gemisinde daha çok hayatta kaldılar ve kayda değer bir frekansa da sahipse bu hiç yoktan bir değişken olarak kullanılabilir. Bununla beraber buradaki hem cinsiyeti hem de olası yaşları ifade edebilecek bu değişkenler de kullanılarak değişken oluşturulabilir. Görülebileceği üzere herhangi bir bilgi taşımayan herhangi bir değişken odaklanıldığında özellik çıkarımı teknikleriyle bunun içerisinden anlamlı olabilecek yeni değişkenler türetilebilir. Burada anlamlı olabilecek ifadesinin altını ilgilenilen problemdeki hedef değişken, bağımlı değişken açısından bir ayırt edicilik oluşturup oluşturmama durumu ifade edilmektedir. 


* Elimizde yuukarıdaki gibi bir metin olsun. Bu metin bir filmin tanımıdır. Diyelim ki doğal dil işleme teknikleriyle ve tavsiye sistemlerinde kullanılan içerik temelli bazı yöntemlerle film öneri sistemi geliştirmek istiyoruz. Burada bir ihtiyaç vardır. Doğal dil işleme ve görüntü işleme problemlerinin kapısındayız şu an. Bazı yapısal ve yapısal olmayan verilerden bahsettik şu ana kadar. Bu metin yapısal olmayan bir veridir. Yani biz bunu bilgisayara nasıl anlatabiliriz ki? Anlatamıyoruz şu anda. Dolayısıyla şöyle bir yyol izleyebiliriz: Örneğin bu film gibi yüzlerce filmin açıklamalarını alırız, bu yüzlerce filmin açıklamalarında yer alan bütün eşsiz kelimeleri sütunlara koyarız. Movie 1'de Word 1'den kaç tane var şeklinde bunları vektörleştirebiliriz. İhtiyaç problem ve problemin çözümü. Burada biz bir ölçüm yapmak, çeşitli tekniklerle bu metnin üzerinde çalışmak istiyorduk ama nasıl çalışacağız bu bir metin? Öyleyse bunu matematiksel bir temsile dökmeliyiz.


* Bunu vektörle temsil edebiliriz. Hepsini temsil ettiğimiz için bir matris formunda bu temsil işlemi gerçekleşmiştir. İhtiyaç matematiksel temsil ihtiyacıdır çünkü üzerinde bazı algoritma vb bazı yöntemler kullanıyor olacağız. Ama kullanamıyoruz çünkü bunlar matematik dünyasında var olan yöntemler. Dolayısıyla bu metni yukarıdaki gibi sayılara dönüştürdüğümüzde üzerinde işlem yapabilir hale geldik. Bu yaptığımız da aslında özellik çıkarımıdır. Özellikler her bir kelimedir. Her bir kelimenin ilgili film açıklamasında kaç defa geçtiği bilgisi özellik çıkarımı sayesinde türetilmiştir.


* Başka bir problemimiz var. Elimizdeki yukarıdaki gibi çeşitli hayvanların görselleri olan bir veri setimiz var. Bu hayvanlarla ilgili tıpki makine öğrenmesi yapıyormuş gibi bir sınıflandırıcı yapmak istiyoruz. Ya da daha basit ifadesiyle bu kedi resmi üzerinde çalışmak istiyoruz. Bunu nasıl yaparız?
* Yine önemli bir farkındalık noktasındayız. Bir problem var. Biz bu resim üzerinde çalışmak istiyoruz ama bu resimi bilgisayar nereden anlayacak. Öyle bir şey yapmalıyız ki bunu lineer cebir dünyasında temsil edilebilir forma getirebilelim, matrisler ve cebirler ile üzerinde çalışabilelim. Dolayısıyla o zaman şu yapılıyor olacak: Buradaki resim bir şekilde matematiksel ifadelerle temsil edilmeli, pixellerin yoğunlukları, belirli bölgelerdeki belirli yoğunluk farklılaşmaları, yine belirli bölgelerdeki pixellerin renk dağılımları gibi bir şekilde buradan feature türetmemiz gerek. Tamamen kafada belirmesi için ifade ediyoruz: Bilgisayar olarak buraya baktığımızda hiçbir şy anlamayız normalde, ama öyle bir şey yapmalıyız ki ifade ettiğimiz gibi kulakla ilgili featurelar dediğimizde kedinin kulağına odaklanıyor oluruz, aynı şekilde kafa, pati için de geçerlidir. Bu resme parça parça odaklanıldığında ve bu parçalara ilişkin bazı featurelar türetildiğinde bu resmi de matematiksel temsillere dönüştürmüş oluruz.

* Deep Learningin de daha doğrusu grörüntü işleme problemlerine uyarlanabileceği Deep Learning çalışmalarının da giriş kapısındayız. 

* Bakın feature map... Bu yapısal olmayan veriyi artık bizim matematiksel işlemlere sokabileceğimiz bir şekilde matematiksel bir forma çevirmiş olmamız durumu söz konusu olur. 
* Özetle makine öğrenmesi, derin öğrenme, hatta zaman serisi problemleri ve bununla ilişliki birçok modelleme sürecinin temelini buradaki kritik eşik noktası yönlendirmektedir. Yani bizim tabular data da olsa resim/görüntü verisi de olsa feature türetmemiz lazım. Bunlar üzerinden featurelar türeterek bunları temsil edebilelim ki buna göre bir sınıflandırma, regresyon, zaman serisi gibi problemleri çözebilelim.
