### Regresyon Modellerinde Başarı Değerlendirme (MSE, RMSE, MAE)

* 
* En son bazı yöntemlerle elde edilen sabit ve ağırlığımız vardı. Bu ölçümlere göre hatalarımıza baktık. Bir ihtiyacımız oluşmuştu bu noktada. Hatalara baktığımızda bazı hataların çok yüksek bazı hataların da çok düşük geldiğini görmüştük. Bunları değerlendiremedik.

* DİKKAT : MSE formülasyonunun neden karesi var sorusunun yanıtındayız çünkü eğer hatalarımızda bir ortalama alırsak artı ve eksiler birbirini götüreceğinden dolayı bu tablo için ortalamamız 0 olabilir. 

* 
* Neden kare alıyoruz? Bunun sebebi ölçüm problemini ortadan kaldırmaktır.

* DİKKAT : Takmamız gereken 2 gözlük var. Birincisi başarı değerlendirme metriği olarak MSE değeri, ikincisi bir optimizasyon problemi olarak MSE değeridir. Bu iki değer ciddi bir farkındalık gerektirmektedir. Buradaki regresyon modellerinin başarısını değerlendiriyoruz şu anda ya b değerlerini nerelere koyacağımızı ağırlıkları nasıl bulacağımızdan girdik ama önce rastgele bazı ağırlıklara sahipmişiz gibi gözüktük. Bunun üzerinden hatalarımızı değerlendiriyoruz.

* Sırası yine bizde aşağıdaki tablodan MSE değerini hesaplayalım...

* 
* CEVAP : 717,08 Ancak tabloda ölçülen hata_kare değerlerine baktığımızda kareden dolayı büyük bir şişme meydana geldiğini görüyoruz.

* 
* Bu şişmeyi RMSE ile geri döndürebiliriz.

* 
* Kare ve kök işlemleri yapmadan mutlak değer işlemi ile de bunu çözebiliriz.

* DİKKAT : Bir modelleme çalışmasında bunlardan hangisini kullanmalıyız sorusu çok kritiktir. Bir önemi yoktur, herhangi birisi tercih edilir ve kullanılır. Birbiri ile kıyaslanıp en düşük çıkanı tercih edelim gibi bir durum yoktur, olmamalıdır. Üçü de kullanılabilir. Biz bu modülde RMSE değerini tercih edeceğiz. Bu rastgele bir tercihtir.
