### Doğrusal Regresyon İçin Gradient Descent (Gradient Descent for Linear Regression)

* Gradient descent makine öğrenmesinden bağımsız olarak bir optimizasyon yöntemidir. Bu optimizasyon yönteminin amacı bir fonksiyonu minimum yapabilecek parametre değerlerini bulmaktır. Bu işlemi şöyle gerçekleştirir: İlgili fonksiyonun ilgili parametreye göre kısmi türevini alır ve bu türev sonucuna göre güncelleme işlemi yapar. Elde edilen türev diğer ifadesiyle gradyan değeri ilgili fonksiyonun maksimum artış yönünü verir. Dolayısıyla ilgili fonksiyonun maksimum artış yönünü veren gradyanın tersine doğru belirli bir şiddet ile giderek parametrenin eski değerinde değişiklik yaparak her iterasyonda hatanın azalmasını sağlar.

###### Peki bu ne demektir?
* Yukarıda grafikte gösterilen sarı noktadan ilerleyerek ağırlığımızı update etmek istediğimizi düşünelim. Bu ağırlığı güncellemek istediğimizde bunun türevini aldık dediğimiz nokta sarı noktadır. Bunun eğimi ise pozitiftir. Hatayı azaltmak için bir yere gitmemizlazım. Eğimi dediğim yer aşağıdaki formüldür:

* 
* Bir fonksiyonun belirli bir noktadaki türevi, gradyanı o fonksiyonun yön bilgisini ve maksimuma gideceği noktayı taşır. Diyelim ki burada bu fonksiyonun bu noktadaki türevi  olsun. Diyelim ki bu ağırlığımız (θ1 değeri) da 8 olsun. Gradyanın tersine gitmek gerek demiştik bu yüzden çıkarma işlemi yapacağız. Yani 8 - 10 olacak ancak 10'un başında bir α sembolü var. α learning rate'tir ve 0.1 olsun diyelim. Bir fonksiyonun belirli bir noktadaki gradyanının tersine gidersek ilgili paramtreyi doğru bir şekilde güncelleyip hatayı azaltacağımız tarafa doğru gideriz dedik. Gradyanı hesapladık, güncelleme için bunun tersine gitmeliyiz diyip çıkarma işlemi yaptık ve belirleyeceğimiz bir hızda gitse dedik, learning rate'i 0.1 olarak aldık. Sonuç olarak θ1 (wi) değerini 7 olarak bulduk. Cost değerimizin 8 ve 7'ye denk gelen kısımlarını grafikte karşılaştırdığımızda yeni θ1 değerinde daha az çıktığını göreceğiz. Şu anda pozitif olan bir gradyana göre sola doğru gitme işini gözlemlemiş olduk.
