### Doğrusal Regresyon (Linear Regression)

* 

* Şapkalı yi ifadesi bağımlı değişkeni ifade ederken xi ifadesi bağımsız değişkenleri ifade etmektedir. wi'ler katsayı veya ağırlıklardır. 
* b ifadesi karşımıza beta, bias, sabit olarak gelebilir.
* wi değişkenleri ilgili değişkenlerin etkilerini (metrekare, apartman yaşı) ifade etmek için kullanılır. Burada + olarak gösterilse de duruma göre eksi veya artı olabilirler. Metrekare arttıkça fiyat da artacaktır bu nedenle w değeri pozitiftir. Apartman yaşı arttıkça fiyat azalacaktır, w değeri negatif olacaktır.
