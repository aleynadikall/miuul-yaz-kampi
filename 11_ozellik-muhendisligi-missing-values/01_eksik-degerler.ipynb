### Eksik Değerler (Missing Values)

* Gözlemlerde eksiklik olması durumunu ifade etmektedir. Diyelim ki elimizde çeşitli değerlerden oluşan aşağıdaki gibi bir veri seti olsun. 

* Bu veri setindeki bazı değerler yerinde yoksa (NA ile ifade edilenler) buunlara eksik değerler denir.
* Eksik veri problemi nasıl çözülür?
* Yaygınca kullanılan 3 ana yönetmi dile getireceğiz.
* 1 - Silme Yaklaşımı
* 2 - Değer Atama Yöntemleri (mod,medyan vb - Bunlara basit atama yöntemleri de denir. Değer atama yöntemleri tahminsel değil ya da gelişmiş değil de daha basit bir şekilde ortalama gibi medyan gibi değerlerle atama işlemlerini ifade etmektedir.)
* 3 - Tahmine Dayalı Yöntemler (Makine öğrenmesi ya da istatiksel bazı yöntemlerle tahminlere göre değer atamaya dayanmaktadır.)
* DİKKAT: Atama işlemlerinin arkasındaki fikir hem çekici hem de tehlikelidir. (R.J.A Little & D.B. Rubin) Peki neden? Eksik veri ile çalışılırken göz önünde bulundurulması gereken bir konu var bu da eksik verilerin rassallığı. Yani eksikliğin rastgele ortaya çıkıp çıkmadığı durumunun bilinmesi gerektiğidir.
* Eksik değere sahip gözlemlerin veri setinden direkt çıkarılması ve rassallığının incelenmemesi, yapılacak istatistiksel çıkarımların ve modelleme çalışmalarının güvenilirliğini düşürecektir. (Alpar)
* Eksik gözlemlerin veri setinden direkt çıkarılabilmesi için veri setindeki eksikliğin bazı durumlarda kısmen bazı durumlarda tamamen raslantısal olarak oluşmuş olması gerekmektedir.
* Eğer eksik değişkenler ile ilişkili olarak ortaya çıkan yapısal problemler ile meydana gelmiş ise bu durumda yapılacak silme işlemleri ciddi yanlılıklara sebep olabilecektir. (tabachnick ve Fidell)
* DİKKAT: Diyelim ki bir değişkeni inceliyoruz, bu değişken şu olsun: Kredi kartı harcama deişkeni olsun. Örneğin bu aylık ortalama ya da kredi kartı harcaması değişkeninde bazı değerlerin NA olduğunu düşünelim. Eğer bu NA'ler rastgele ortaya çıktıysa problem yok, ister sileriz ister basit atama yöntemleri yaparız ister tahmine dayalı yöntem kullanırız. Peki yukarıdaki alıntılarla dile getirdiğimiz ifade ne demek? Şimdi aynı senaryo için yani kredi kartı harcaması değişkeninin olduğu bu senaryo için varsayalım ki kredi kartı var mı yok mu değişkeni olsun. Bir kişinin kredi kartı yoksa yani 0'sa zaten kredi kartı harcaması 0 olacaktır. 0 numerik olarak basılmamış olacağından dolayı da 0 olması yerine NA olacaktır. Dolayısıyla ee hani bu değişken rastgele olmalıydı demiştik ya? Rastgele değil işte. Çünkü kredi kartı olup olmama durumunu ifade eden başka bir değişkene bağlı. İşte buna bir değişkendeki eksiklik başka bir değişken etkisinde ortaya çıkmıştır denir. Bundan dolayı kredi kartı harcaması değişkenindeki eksiklikler rastgele değildir. Eee biz bunu ne yapacağız? Diyelim ki çeşitli yöntemler kullandığımızda görüyoruz ki eksik değerler rastgele değil bunun sebebi ne olabilir? İşte başka bir değişkenle arasında bağımlılık olabilir. Bundan dolayı ortaya çıkmış olabilir. İşte bu durumu biliyor olacağız. Eğer eksiklikleri doldurmaya karar verirsek ve eksikliklerin de rastgele oluşmadığını bilirsek, bu durumda yapacağımız işlemlerin ortaya bazı problemler çıkarabileceğini varsaymamız gerekiyor. Eğer eksiklikler rastgele ise rahatız, eğer eksiklikler rastgele ortaya çıkmadıysa rahat değiliz. Adım atarken birkaç defa düşünmemiz gerekir.
