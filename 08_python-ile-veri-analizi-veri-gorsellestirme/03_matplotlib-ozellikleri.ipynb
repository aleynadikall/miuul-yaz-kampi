### Matplotlib'in Özellikleri

```python
# Yapısı itibari ile katmanlı şekilde veri görselleştirme imkanı sağlar.

# Bu şu anlama gelir: bir katmanda bir görsel diğer katmanda ayrı bir görsel  diğer katmanda bir title bir isimlendirme 
# örneğin diğer bir katmanda ise eksenlere bilgi vermek gibi diğer çeşitli bazı başlıklarda çalışma imkanı sağlar.

# Bu yapısını kullanarak bazı ihtiyaçlarımızı giderebiliriz.

# Pratikte uygulama esnasında ya da kod okumaları yaparken çeşitli kaynaklardan blok okumaları yaparken karşımıza gelebilecek 
# bazı matplotlib özelliklerini ifade edeceğiz.

# Bizim için önemi ne ne kadar odaklanmalıyız?: Sadece biliyor olmak yeterli olacaktır. Bu noktada böyle de bir konu varmış 
# diye bir bilgimiz olması yeterli olacaktır.

# Veri görselleştirmeye bakışımız daha ölçeklenebilir daha büyük çaplı işler içindir. Öyle ki fonksiyonel keşifçi veri analizi 
# bölümünde buradan sadece örneğin value_count fonksiyonunu ve histogram fonksiyonunu alıp yüzlerce ya da binlerce değişken 
# olduğunda burada nasıl ölçeklenebilir bir veri analizi süreci gerçekleştirebiliriz bunu konuşacağız. Dolayısıyla burada bu 
# özellikleri bir bir bilmek işimize yarıyor olacaktır. Ana odağımız eeen derinlerde detaylarda boğulmak yerine daha yüksek 
# seviyeden büyük ölçekli işler yapmak için ihtiyaçlarımızı görebilmektir.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
```

### plot özelliği

```python
x = np.array([1, 8])
y = np.array([0, 150])
```

```python
plt.plot(x, y)
plt.show()
```

```python
# Çizgisiz şekilde:
plt.plot(x, y, 'o')
plt.show()
# Vahit hoca bu kısımda pcharm kullandığı için ilk görselleştirme kodunu çalıştırdıktan sonra çıkan ekranı kapatmadan bu 
# kısımdaki görselleştirme kodunu çalıştırdığından olan değişiklik ilk çıktının üzerinde olacak şekilde bize gösterildi.
# Doğru bir sonuç elde etmek istiyorsak ekranı kapatıp kodu öyle çalıştırmalıyız. Eğer pycharm'da çalışırsan buna dikkat
# etmelisin.
```

```python
# Daha fazla nokta olursa bunu nasıl yapar?
x = np.array([2, 4, 6, 8, 10])
y = np.array([1, 3, 5, 7, 9])
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

### marker özelliği

```python
# İşaretleyici / işaret özelliği
y = np.array([13, 28, 11, 100])
```

```python
# y noktalarını istediğimiz marker ile işaretlemesi amacındayız:
plt.plot(y, marker = 'o')
plt.show()
```

```python
# farklı marker şekli ile de işaretleyebilirdik:
plt.plot(y, marker = '*')
plt.show()
```

```python
# çeşitli marker işaretleri vardır, bunlar:
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h']
```

### line özelliği

```python
np.array([13, 28, 11, 100])
plt.plot(y)
plt.show()
```

```python
# bir farkındalık olması açısından argümanını değiştirmek isteyebiliriz:
# Kesikli ve noktalı
np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashdot", color="r")
plt.show()
```

```python
# noktalı halde de yapabiliriz.
np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dotted")
plt.show()
```

```python
# Kesikli
np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashed", color="b")
plt.show()
```

### Multiple Lines

```python
# Matplotlib'nin katmanlı yapısından dolayı çok rahat kullanabiliyoruz.
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
```

* DİKKAT: Şu anda biz bunu bu editörde kullanırken (pycharm) plot.show'a gerek kalmaksızın ekrana bastırdığını görüyoruz. Bunun sebebi yeni matplotlib sürümüdür. Bu kullanımdan ziyade mutlaka show ifadesini kullanmaya alışkanlık kazanılması önerilir. Herhangi başka editörde kullanırken de plot.show()'u alışkanlık haline getirmeniz önerilir.

### Labels özelliği

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
```

```python
# Garfiğe başlık eklemek istediğimizi düşünelim:
plt.title("Bu ana başlık")
plt.plot(x, y)
```

```python
# X ekseninde isimlendirme yapmak istersek:
plt.xlabel("X ekseni isimlendirmesi")
plt.plot(x, y)
```

```python
# Y eksenini isimlendirme
plt.ylabel("Y ekseni isimlendirmesi")
plt.plot(x, y)
```

```python
# Izgara ekleyelim:
plt.grid()
plt.plot(x, y)
plt.show()
```
