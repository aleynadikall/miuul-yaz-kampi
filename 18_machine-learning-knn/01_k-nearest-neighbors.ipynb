### K-En Yakn Komu ( K-Nearest Neighbors )

#### K-En Yakn Komu Regresyon (K-Nearest Neighbors Regression)

* 

* Bana arkadan syle sana kim olduunu syleyeyim eklinde alr. KNN yntemi bir gzlem biriminin kendine en yakn olan dier gzlem birimleri hesaplanr. Bir gzlem biriminin kendisine en yakn olan dier k adet gzlem birimi hesaplanr ve bu dier kendisine en yakn k adet gzlem biriminin baml deikeni neyse bunun zerinden ilgili deiken iin tahminde bulunulmu olunur.

* 

* Baml deikenimiz saysal olduu iin anlyoruz ki bu bir regresyon problemidir. KNN regresyon ve snflandrma modelleri iin kullanlabilir. Bu noktada bir tahmin etme ilemi gerekletirmek istediimizi dnelim.
* x1 = 50, x2 = 230 eklinde birgzlem birimimiz olsun diyelim. Problemimiz u: Bu bamsz deiken deerlerine sahip bir gzlem biriminin baml deiken deeri nedir. Y'nin tahmini deeri nedir eklinde bir sorumuz var.

* Bu noktada yle bir ilem gerekletirmemiz grekiyor. klit ya da benzeri bir uzaklk yntemi ile bu gzlem biriminin (x1 = 50, x2 = 230) her bir gzlem birimiyle arasndaki uzakl hesaplayp bu uzakl hesapladktan sonra rnein kendisine en yakn k tane gzlem birimine baml deikenin ortalamasn almamz gerekiyor. 

* 

* Siyah nokta bizim gzlem birimimizdir.

* 

* 50 - 56 ^ 2 + 230 - 241 ^ 2 eklinde son x1 x2 deerine kadar bu ilemi yapar ve elde edilen saysal deerlerden en kk olan bu gzlem birimine en yakn olandr deerlendirmesi yaparz. Dolaysyla bu gzlem birimine en yakn 5 tane gzlem biriminin baml deikeninin ortalam deerini hesapladmzda bu durumda bu gzlem birimi iin tahmin sonucunu ortaya karm oluyoruz.
