### K-En Yakın Komşu ( K-Nearest Neighbors )

#### K-En Yakın Komşu Regresyon (K-Nearest Neighbors Regression)

* 

* Bana arkadaşını söyle sana kim olduğunu söyleyeyim şeklinde çalışır. KNN yöntemi bir gözlem biriminin kendine en yakın olan diğer gözlem birimleri hesaplanır. Bir gözlem biriminin kendisine en yakın olan diğer k adet gözlem birimi hesaplanır ve bu diğer kendisine en yakın k adet gözlem biriminin bağımlı değişkeni neyse bunun üzerinden ilgili değişken için tahminde bulunulmuş olunur.

* 

* Bağımlı değişkenimiz sayısal olduğu için anlyoruz ki bu bir regresyon problemidir. KNN regresyon ve sınıflandırma modelleri için kullanılabilir. Bu noktada bir tahmin etme işlemi gerçekleştirmek istediğimizi düşünelim.
* x1 = 50, x2 = 230 şeklinde birgözlem birimimiz olsun diyelim. Problemimiz şu: Bu bağımsız değişken değerlerine sahip bir gözlem biriminin bağımlı değişken değeri nedir. Y'nin tahmini değeri nedir şeklinde bir sorumuz var.

* Bu noktada şöyle bir işlem gerçekleştirmemiz grekiyor. Öklit ya da benzeri bir uzaklık yöntemi ile bu gözlem biriminin (x1 = 50, x2 = 230) her bir gözlem birimiyle arasındaki uzaklığı hesaplayıp bu uzaklığı hesapladıktan sonra örneğin kendisine en yakın k tane gözlem birimine bağımlı değişkenin ortalamasını almamız gerekiyor. 

* 

* Siyah nokta bizim gözlem birimimizdir.

* 

* 50 - 56 ^ 2 + 230 - 241 ^ 2 şeklinde son x1 x2 değerine kadar bu işlemi yapar ve elde edilen sayısal değerlerden en küçük olan bu gözlem birimine en yakın olandır değerlendirmesi yaparız. Dolayısıyla bu gözlem birimine en yakın 5 tane gözlem biriminin bağımlı değişkeninin ortalam değerini hesapladığımızda bu durumda bu gözlem birimi için tahmin sonucunu ortaya çıkarmış oluyoruz.
