### ROC Eğrisi ( ROC Curve )

* 

* True positive rate y ekseninde, false positive rate x eksenindedir. Eğer hiçbir model kurmasaydık ve rastgele 1 0 atamaları yapsaydık, bu durumda elde edeceğimiz başarı kesikli kırmızı çizgi olacaktı. Yani burası bir temel karşılaştırma noktasıdır. Bu eksen üzerindeki yeşil, turuncu, mavi çizgiler yukarıya doğru yani kapladıkları alan arttıkça başarı oranı artıyor olarak düşünülür. True positive rate PRECISION ifadesine denk gelirken false positive rate (Yanlış alarm üretme oranı)'tir. Dolayısıyla eksenlerimiz bunlardır. Anlıyoruz ki ne kadar yukarıda o kadar iyidir. 

###### Peki bu eğriler nasıl ortaya çıktı?

* DİKKAT : Az önce altyapısını hazırlamıştık. Threshold değerleri 0 ile 1 arasında çeşitli değerler hazırlasın ve bunlardan bir tanesini (threshold değerini) 0.1, diğerini 0.5, başka bir diğerini 0.75 yaptık. Diyelim ki thresholdu da 0.1 yaptık. Thresholdu 0.1 yaptığımızda tüm gözlem birimlerimize ilişkin tahminlerimizi oluşturuyoruz (Birisine 1 birisine 0 diğerine 0 bir başkasına 1 şeklinde). Bunun üzerinden karmaşıklık matrisini oluşturuyoruz. Bu karmaşıklık matrisini oluşturduktan sonra True Positive Rate ve False Positive Rate'e bakıp bunların karşılık geldiği noktayı işaretliyoruz. Sonra eşik değeri 0.5 olarak değiştiriyoruz. Eşik değeri 0.5 yaptıktan sonra tekrar 1 0 1 0 sınıf atamaları yapıp bunun üzerinden karmaşıklık matrisini oluşturuyoruz. Bu matrisi oluşturduktan sonra tekrar True Positive Rate ve False Positive Rate'e bakıyoruz. Karşılıklarını hesaplıyoruz. Benzer şekilde tüm thresholdlara baktığımızda ROC Eğrisi çıkmaktadır. Threshold değiştiğinde tüm metrik değerleri değişiyor demiştik. Öyleyse biz bütün olası thresholdları çıkaralım ve bu olası thresholdlara göre karmaşıklık matrisini oluşturalım ve bunun üzerinden her bir olası eşik değerine karşılık True Positive Rate ve False Positive Rate'i hesaplayıp grafikte işaretleyelim. Eğri ne kadar yer kaplıyorsa model o kadar başarılıdır. 

###### Peki bu eğrileri nasıl değerlendirmeliyiz?

* ROC Eğrisinin altında kalan alanın bir integrali alınırsa Area Under Curve (AUC) elde edilir. AUC metriği ROC Curve'ün sayısal bir şekilde ifade edilişidir.
