### Sınıflandırma Problemlerinde Başarı Değerlendirme

* Diyelim ki bir müşteri terk modeli geliştirme sürecindeyiz. Bizi terk eden müşteriler 1 ile ifade edilmiş, terk etmeyenler 0 ile ifade edilmiş. Öyle bir model kurmak istiyoruz ki müşterilerimizin bizi terk edip etmeyeceğini tahmin etmeye çalışalım. Herhangi bir teknik kullanılmış olsun. Lojistik regresyon olabilir, ağaca dayalı yöntemler, yapay sinir ağları olabilir. Herhangi bir makine öğrenmesi yöntemi ile bizim elimizdeki veriden öğrendiğimiz yapıyı kullanarak tahminlerimiz olsun. Burada başarı 1 olana 1 demektir. 0 olana 0 demektir. Dolayısıyla aşağıda doğru yaptığımız tahminler olduğı gibi yanlış yaptığımız tahminler de mevcuttur. 

* 
* 

* Burada doğruların oranını bulmak istersek doğrular/tamamı diyerek 7/10 sonucunu buluruz. Buna accuracy score (doğruluk oranı) denir. Temelinde bununla ilintili bazı diğer metrikleri görüyor olacağız.

###### Confussion Matrix (Karmaşıklık Matrisi)

* Elimizdeki veri setlerinde henüz modelleme aşamasına geçmeden önce ilgilendiğimiz probleme göre 1 ve 0 sınıfları vardır. Daha sonra model kurarak bizim tahmin edeceğimiz sınıflar var. Bazen 0 için 0 veya 1 için 1 tahminini yaparken bazense 1 için 0 veya 0 için 1 olacak şekilde doğru ya da yanlış tahmin ve gerçek tahmin ikilileri elde edebiliriz. 1'e 1 - 0'a 0 dediğimiz noktalarda başarılı, 1'e 0 - 0'a 1 dediğimiz noktalarda başarısız işlemler yapmış olabiliriz.

* 

* Doğru yaptığımız işlemlerin başında T(True) vardır. 1 ifadesi pozitif sınıfı temsil eder, bu nedenle P şeklinde yazılır.

* 

* Diyelim ki bir sahtekarlık tespiti çalışması üzerinde çalışıyoruz. Sahtekar olarak ifade edilen sınıf 1 olsun. Bu durumda bizim 1 olarak tahmin ettiklerimiz sahtekar sınıfı olacaktır. Precision'a göre sahtekar olarak tahmin edilenlerin gerçekte ne kadarının sahtekar olduğunu ifade ederiz. Bu metrikte bir işlem sahtekarlık işlemi olmadığı halde sahtekarlık işlemi gibi muamele görebilir. Bunun karşılığı FP'dir. Bu durumda yanlış alarm üretmiş oluruz. FP yanlış alarm üretmek demektir, ayrıca istatistikte 1.tip hata da denir. 

* 

* Recall pozitif sınıfın doğru tahmin edilme oranıdır. Gerçekte belirli sayıda bir pozitif sınıf var, burada ilgilendiğimiz senaryo kapsamında sahtekarlık işlemleridir. Bu gerçekte sahtekarlık işlemi olan işlemlerin ne kadarını doğru tahmin ettiğimizi gösterir. Sahtekarlık işlemlerinin ne kadarının doğru tahmin edildiğini ifade eder. Bazı sahtekarlık işlemleri var (ilk satırda). Ama sahtekar olduğu halde sahtekar değil dediklerimiz de var. Yani sahtekar olduğu halde yakalayamadığımız bazı işlemler var (FN). Bu 2.tip hatadır, kritik hata denir çünkü sonucunda ciddi sorunlar ortaya çıkabilir.

* 

###### Peki bu bizim ne işimize yarayacak?

* Eğer elimizdeki sınıflandırma problemi dengeli sınıf dağılımına sahipse accuracy'i verebiliriz. Ama eğer elimizdeki veri setinde sınıflar dengesiz dağılıyorsa bu durumda direkt accuracy oranını kullanamayız. Recall ve precision değerlerine bakmamız gerekir. 

* 

* F1 Score'u precision ve recall değerlerinin harmonik ortalamasıdır. Precision değeri tahminlerimizin başarısına odaklanmıştır. Recall değeri gerçekleri yakalama oranımıza odaklanmıştır. Dolayısıyla bizim için önemlidir. Bu açıdan F1 Score değeri ikisinin de etkisini tutmaktadır.
