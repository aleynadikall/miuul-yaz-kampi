###### Confussion Matrix (Karmaklk Matrisi)

* Elimizdeki veri setlerinde henz modelleme aamasna gemeden nce ilgilendiimiz probleme gre 1 ve 0 snflar vardr. Daha sonra model kurarak bizim tahmin edeceimiz snflar var. * Doru yaptmz ilemlerin banda T(True) vardr. 1 ifadesi pozitif snf temsil eder, bu nedenle P eklinde yazlr.

* Diyelim ki bir sahtekarlk tespiti almas zerinde alyoruz. Sahtekar olarak ifade edilen snf 1 olsun. Bu durumda bizim 1 olarak tahmin ettiklerimiz sahtekar snf olacaktr. Precision'a gre sahtekar olarak tahmin edilenlerin gerekte ne kadarnn sahtekar olduunu ifade ederiz. Bu metrikte bir ilem sahtekarlk ilemi olmad halde sahtekarlk ilemi gibi muamele grebilir. Bunun karl FP'dir. Bu durumda yanl alarm retmi oluruz. FP yanl alarm retmek demektir, ayrca istatistikte 1.tip hata da denir.

* Recall pozitif snfn doru tahmin edilme orandr. Gerekte belirli sayda bir pozitif snf var, burada ilgilendiimiz senaryo kapsamnda sahtekarlk ilemleridir. Bu gerekte sahtekarlk ilemi olan ilemlerin ne kadarn doru tahmin ettiimizi gsterir. Sahtekarlk ilemlerinin ne kadarnn doru tahmin edildiini ifade eder. Baz sahtekarlk ilemleri var (ilk satrda). Ama sahtekar olduu halde sahtekar deil dediklerimiz de var. Yani sahtekar olduu halde yakalayamadmz baz ilemler var (FN). * F1 Score'u precision ve recall deerlerinin harmonik ortalamasdr. Precision deeri tahminlerimizin baarsna odaklanmtr. Recall deeri gerekleri yakalama oranmza odaklanmtr. Dolaysyla bizim iin nemlidir. Bu adan F1 Score deeri ikisinin de etkisini tutmaktadr.
