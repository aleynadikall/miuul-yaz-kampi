### Lojistik Regresyon (Logistic Regression)

* 

* DİKKAT : Normalde bağımlı değişkenin 1 ve 0'lardan oluştuğunu düşünürsek bağımsız değişkenlerin de çeşitli değerlerden oluştuğunu düşünelim. Normalde bir regresyon modeli kurduğumuzda bağımlı değişken 1 ve 0 olduğunda bu modeli kurabiliriz, bir problem yok. Fakat bu modeli kurduğumuzda tahmin sonuçlarını elde etmek istediğimizde değerler 1'den büyük veya 0'dan küçük çıkabilir, eksili değerler çıkabilir. Dolayısıyla öyle bir işlem yapmamız lazım ki tahmin edilen nihai sonuç 0-1 arasında yer alsın ve bunu belirli bir eşik değerine göre 0 ya da 1 olarak değerlendirebilelim. İşte bu amacı yerine getirebilmek için yukarıdaki gibi bir dönüştürme işlemi yapılır. Kesirli ifadeye sigmoid fonksiyonu denir. Bulunan z değerini sigmoid fonksiyonunda yerine yazdığımızda bize 0 ile 1 arasında bir olasılık değeri üretilir. Bu olasılık değerine göre 0 ya da 1 sınıfı ataması yapılır. 

* 

###### Sigmoid Fonksiyonu Nedir ve Dönüştürme İşlemi Nasıl Gerçekleştirilir?

* 

* Yukarıdaki fonksiyon sigmoid fonksiyonudur. Buradaki x ekseninin ifade ettiği bu doğrusal formdan (b + w1x1 + w2x2 + ... + wpxp) gelecek olan skorlardır/değerlerdir. Belirli bir yöntemle ağırlıkları bulduğumuzu varsayacak olursak bu ağırlıkları yerine koyduğumuzda belirli bir gözlem biriminin değerleri geldiğinde yani bağımısz değişken değerleri geldiğinde bu etkilişimden sonra çıkan değer z değeridir. Buradan gelen z değeri sigmoid fonksiyonuna (Kesirli formül) sokulur. Bu sigmoid fonksiyonunun karşılığı ise grafikteki mavi çizgidir. Bu fonksiyondan çıkacak olan değer 0 ve 1 arasında olacaktır. Yani klasik regresyon yapsaydık ve bağımlı değişken 1 ve 0'lardan oluşsaydı bu durumda elde edeceğimiz değerlerin 1 ve 0 arasında olması durumu garanti edilemiyor gibi bir durum vardı. Eee öyleysee biz öyşe bir fonksiyon kullanalım ki bunun çıktısını (z değerini) 0 ile 1 arasına eşlesin. Sigmoid fonksiyonunun görevi z değerini 0 ve 1 arasında bir değere dönüştürmektir.

* DİKKAT : y eksenine yani tahmin edilen değerlere geliyoruz. x bağımsız değişkenleri değerlerini aldığında y bağımlı değişkenin 1.sınıfının gerçekleşmesi olasılığı ile ilgileniriz. Yani sigmoid fonksiyonunda bu dönüşüm işlemi yapıldığında ilgilendiğimiz problemde 1 sınıfı her neyse (hayatta kaldı, kalamadı, hasta, hasta değil gibi) bu fonksiyonun (yi fonksiyonu) ürettiği değer 0.78 gibi bir olasılık değeridir. Dolayısıyla sınıflandırma problemleri ile ilgilenirken aslında elimizde yapacak olduğumuz tahmin sonuçları neticesinde 1 sınıfına ait olma olasılığı olur. Daha sonra bu 1 sınıfına ait olma olasılığını bir eşik değere göre dönüştürürüz.
