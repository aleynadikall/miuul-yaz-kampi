* Accuracy'e baktığımızda başarılı gibi gözüken model recall ve precision'a baktığımızda o kadar da başarılı gözükmüyor. Yine dikkatimizi çekmelidir ki precision değeri ana odağımızda olan 1 sınıfıyla ilgili yaptığımız tahminlere göre oluşturulmuşken recall değeri 1 sınıfını ne kadar doğru tahmin ettiğimizle ilgilenir.
