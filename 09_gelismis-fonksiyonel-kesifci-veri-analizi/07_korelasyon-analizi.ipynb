### Korelasyon Analizi (Analysis of Correlation)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = pd.read_csv("breast_cancer.csv")
df.head()
```

```python
# Baştaki id ve sondaki unnamed:32 değişkenleri istemediğimiz değişkenler, onlardan kurtulalım:
df = df.iloc[:, 1:-1]
df.head()
```

* AMAÇ: Elimize bir veri seti geldiğinde bunun ısı haritası aracılığıyla korelasyonlarına bakmak ve daha sonra yüksek korelasyonlu bir değişken setindeki yüksek korelasyonlu değişkenlerden bazılarını dışarıda bırakabilmeyi görmektir.
* DİKKAT: Burada görecek olduğumuz yüksek korelasyonlu değişkenlerden birisini silme konusu illa ki her çalışmada yapılacaktır diye bir durum yoktur. Sadece ihtiyacımız olduğunda başvurmamız gerekir.

```python
# Numerik değişkenleri seçecek basit bir list comprehension yazalım:
num_cols = [col for col in df.columns if df[col].dtype in [int, float]]
```

```python
# Korelasyonları hesaplamak için corr fonksiyonu kullanılır:
corr = df[num_cols].corr()
```

```python
corr
```

#### Korelasyon Ne Demektir?
* Korelasyon değişkenlerin birbiriyle ilişkisini ifade eden bir istatistiksel ölçümdür. -1 ile +1 arasında değerleri alır. -1 veya +1'e yaklaştıkça ilişkinin şiddeti kuvvetlenir. Eğer iki değişkenin arasındaki ilişki pozitifse buna pozitif korelasyon denir ve bir değişkenin değeri arttıkça diğerinin de değeri artar. İki değişken arasındaki ilişki negatifse, bir değişkenin değeri artarken diğer değişkenin değeri azalır. Bu korelasyonlarda ifade edildiği üzere -1 vrya +1'e yaklaştıkça ilişki şiddeti kuvvetlidir.0 civarındaki bir korelasyon değeri varsa korelasyon olmadığı anlamına gelir. Genelde analitik çalışmalarda birbiriyle yüksek korelasyon ilişkisine sahip değişkenlerin çalışmalarda bulunmamasını isteriz. Çünkü ikisi de aynı şeyi ifade ediyor. Yani iki tane yüksek korelasyonlu değişkenin %99 korelasyona sahip olması bu ikisinin nerdeyse aynı değişkenler olduğu anlamına gelir. Bu sebeple birçok çalışmada bu değişkenlerden birisini çalışmanın dışında bırakmak isteriz. 

```python
# Hızlı bir şekilde ısı haritası oluşturalım:
sns.set(rc={'figure.figsize': (12, 12)})
sns.heatmap(corr, cmap="RdBu")
plt.show()
```

#### Yüksek Korelasyonlu Değişkenlerin Silinmesi

```python
# Korelasyonun negatif veya pozitif olmasıyla ilgilenmediğimizden dolayı hepsinin mutlaka değerini alıp pozitif hale getiriyoruz:
cor_matrix = df.corr().abs()
```

```python
# DİKKAT: Erişmek istediğimiz çıktı şu olacak:
#           0         1         2         3
# 0  1.000000  0.117570  0.871754  0.817941
# 1  0.117570  1.000000  0.428440  0.366126
# 2  0.871754  0.428440  1.000000  0.962865
# 3  0.817941  0.366126  0.962865  1.000000

# Yukarıdaki matriste tekrarlayan gereksiz elemanlar olduğu göze çarpıyor. Bunu düzenleyerek aşağıdaki şekle dönüştüreceğiz:
# Böyle bu matristeki tüm gereksiz bilgilerden kurtulmuş olacağız.

#     0        1         2         3
# 0 NaN  0.11757  0.871754  0.817941
# 1 NaN      NaN  0.428440  0.366126
# 2 NaN      NaN       NaN  0.962865
# 3 NaN      NaN       NaN       NaN
```

```python
# 1'lerden oluşan oluşturduğumuz matrisin boyutunda bir numpy array oluşturmakla işe başlıyoruz.
# Sonrasında bu numpy array'ini boola çeviriyoruz. (True False değerleri içerecek formata)
# Daha sonra yukarıda görmüş olduğumuz yapıya çevirmek için numpy'daki ilgili fonksiyonu kullanıyoruz:
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
upper_triangle_matrix
# Aşağıdaki şekilde NaN'ler elde etmiş olduk ve o çoklama durumundan kurtulmuş olduk.
```

* Yukarıdaki çıktıda radius_mean ve perimeter_mean değişkenlerinin korelasyonu 0.997855'tir. Bu nedenle bunu silmememiz gerekiyor. Aynı oldukları anlamına geliyor. Şu an buna ihtiyacımız olduğu için kullanacağız ancak her çalışmada ihtiyaç duyulmaz. Buna dikkat edelim.

```python
drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col]>0.90) ]
drop_list
```

```python
# Artık veri setinden silebiliriz:
df.drop(drop_list, axis = 1)
```

```python
# Bu işlemi fonksiyonlaştıralım:
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={'figure.figsize': (15, 15)})
        sns.heatmap(corr, cmap="RdBu")
        plt.show()
    return drop_list
```

```python
high_correlated_cols(df)
```

```python
drop_list = high_correlated_cols(df, plot=True)
```

```python
df.drop(drop_list, axis=1)
```

```python
high_correlated_cols(df.drop(drop_list, axis=1), plot=True)
```
